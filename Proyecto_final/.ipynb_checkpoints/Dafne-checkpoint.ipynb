```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import networkx as nx
from mpl_toolkits.mplot3d import Axes3D
import scipy.io
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score
from scipy.spatial.distance import pdist, squareform

# Cargar los datos
mat_path = r'C:\Users\daphn\Documents\UNAM\Neurociencias\Quinto semestre 1\Modelos\Git\Neurociencias-2025-1\BCT\BCT\2019_03_03_BCT\data_and_demos\Coactivation_matrix.mat'
mat_data = scipy.io.loadmat(mat_path)

# Obtener la matriz de coactivaci√≥n y las coordenadas
coactivation_matrix = mat_data['Coactivation_matrix']
coordinates = mat_data['Coord']
```

# Ejercicio 1

```python
# Umbrales para los grafos
thresholds = [0.08, 0.09, 0.1]

# Crear los grafos para los tres umbrales
fig = plt.figure(figsize=(18, 6))
for i, threshold in enumerate(thresholds):
    # Crear matriz de adyacencia
    adjacency_matrix = (coactivation_matrix > threshold).astype(int)

    # Crear el grafo
    G = nx.from_numpy_array(adjacency_matrix)  
    degree = dict(G.degree())

    # Coordenadas de los nodos
    pos_3d = {i: coordinates[i] for i in range(len(G))}
    x = [pos[0] for pos in pos_3d.values()]
    y = [pos[1] for pos in pos_3d.values()]
    z = [pos[2] for pos in pos_3d.values()]

    # Crear la subgr√°fica
    ax = fig.add_subplot(1, 3, i + 1, projection='3d')

    # Dibujar nodos
    ax.scatter(x, y, z, s=[degree[node] * 10 for node in G.nodes],
               c=list(degree.values()), cmap='autumn', label='Nodos')

    # Dibujar bordes
    for edge in G.edges:
        start, end = edge
        ax.plot([x[start], x[end]], [y[start], y[end]], [z[start], z[end]],
                color='gray', linewidth=0.5)

    ax.set_title(f'Umbral > {threshold}')
    ax.view_init(elev=20, azim=80)

plt.tight_layout()
plt.show()
```

# Ejercicio 2

```python
# Grafo con umbral 0.09 y animaci√≥n 3D
threshold = 0.09
adjacency_matrix = (coactivation_matrix > threshold).astype(int)
G = nx.from_numpy_array(adjacency_matrix)  
degree = dict(G.degree())

# Coordenadas 3D para los nodos
pos_3d = {i: coordinates[i] for i in range(len(G))}
x = [pos[0] for pos in pos_3d.values()]
y = [pos[1] for pos in pos_3d.values()]
z = [pos[2] for pos in pos_3d.values()]

# Crear figura para animaci√≥n
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


# Inicializaci√≥n de los datos en el gr√°fico
def init():
    """Inicializar el gr√°fico."""
    ax.clear()
    ax.scatter(x, y, z, s=[degree[node] * 10 for node in G.nodes], c=list(degree.values()), cmap='autumn')
    for edge in G.edges:
        start, end = edge
        ax.plot([x[start], x[end]], [y[start], y[end]], [z[start], z[end]],
                color='gray', linewidth=0.5)
    ax.set_xlim(min(x), max(x))
    ax.set_ylim(min(y), max(y))
    ax.set_zlim(min(z), max(z))
    ax.set_title('Animaci√≥n del Grafo (360¬∞)')
    return ax,


# Funci√≥n de actualizaci√≥n
def update(angle):
    """Actualizar el gr√°fico para la animaci√≥n."""
    ax.view_init(elev=20, azim=angle)
    return ax,


# Crear la animaci√≥n
anim = FuncAnimation(fig, update, init_func=init, frames=np.arange(0, 360, 2), interval=100)

# Guardar la animaci√≥n como GIF
anim.save('grafo_rotacion.gif', writer='pillow')

plt.show()
```

```python
# Identificar hubs (nodos con grado > 9)
hub_nodes = [node for node, deg in degree.items() if deg > 9]
hub_subgraph = G.subgraph(hub_nodes)

# Crear la figura del subgrafo de hubs
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Coordenadas de los hubs
hub_pos = {i: coordinates[i] for i in hub_nodes}
hub_x = [hub_pos[node][0] for node in hub_nodes]
hub_y = [hub_pos[node][1] for node in hub_nodes]
hub_z = [hub_pos[node][2] for node in hub_nodes]

# Dibujar nodos de hubs
ax.scatter(hub_x, hub_y, hub_z, s=[degree[node] * 10 for node in hub_nodes], c='red', label='Hubs')

# Dibujar bordes entre hubs
for edge in hub_subgraph.edges:
    start, end = edge
    ax.plot([hub_pos[start][0], hub_pos[end][0]],
            [hub_pos[start][1], hub_pos[end][1]],
            [hub_pos[start][2], hub_pos[end][2]], color='black', linewidth=1.5)

# Ajustar la vista
ax.set_title('Subgrafo de Hubs (Grado > 9)')
ax.view_init(elev=20, azim=80)
plt.show()
```

# Ejercicio 4

```python
# Ejercicio 4

# Matriz de emparejamiento (correlaci√≥n de la matriz de adyacencia)
Mat_Emp = np.corrcoef(mat_data['Coactivation_matrix'])

# Evaluar diferentes valores de K
k_range = range(2, 11)  # Rango de m√≥dulos (K) a evaluar
scores = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(Mat_Emp)
    score = calinski_harabasz_score(Mat_Emp, labels)
    scores.append(score)

# Encontrar el K con el mejor puntaje
optimal_k = k_range[np.argmax(scores)]

print(f'N√∫mero √≥ptimo de m√≥dulos (K) basado en Calinski-Harabasz: {optimal_k}')

# Visualizar la mejor partici√≥n
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
final_labels = kmeans.fit_predict(Mat_Emp)

# Mostrar los resultados de la partici√≥n
print(f'Partici√≥n de nodos en {optimal_k} m√≥dulos:', final_labels)

# Graficar la puntuaci√≥n de Calinski-Harabasz para cada K
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(k_range, scores, marker='o', linestyle='-', color='b')
plt.title('Evaluaci√≥n del N√∫mero de M√≥dulos (Calinski-Harabasz)')
plt.xlabel('N√∫mero de M√≥dulos (K)')
plt.ylabel('Puntuaci√≥n Calinski-Harabasz')
plt.grid()
plt.show()

# A partir de la m√©trica de Calinski-Harabasz, que mide la compacidad y separaci√≥n de los cl√∫steres y el gr√°fico
# visualizado a partir de los c√°lculos, la mejor partici√≥n es en 4 m√≥dulos seguida de 3.
```

# Ejercicio 6

```python
# Obtener la matriz de coactivaci√≥n
coactivation_matrix = mat_data['Coactivation_matrix']

# Asegurar que la matriz sea sim√©trica y sin autoconexiones
coactivation_matrix = (coactivation_matrix + coactivation_matrix.T) / 2
np.fill_diagonal(coactivation_matrix, 0)

# Crear el grafo desde la matriz de coactivaci√≥n
G = nx.from_numpy_array(coactivation_matrix)

# Eliminar nodos aislados para evitar errores en c√°lculos
G.remove_nodes_from(list(nx.isolates(G)))

# Manejo del c√°lculo del coeficiente de Rich Club para evitar divisiones por cero
def safe_rich_club_coefficient(graph, normalized=True):
    """Calcula el coeficiente de Rich Club evitando divisiones por cero."""
    rich_club = nx.rich_club_coefficient(graph, normalized=normalized)
    # Filtrar cualquier entrada donde el coeficiente sea indefinido
    rich_club_safe = {k: v for k, v in rich_club.items() if v is not None and not np.isnan(v) and not np.isinf(v)}
    return rich_club_safe

# Calcular el coeficiente de Rich Club (sin normalizar)
rich_club_unnormalized = safe_rich_club_coefficient(G, normalized=False)

# Calcular el coeficiente de Rich Club (normalizado)
rich_club_normalized = safe_rich_club_coefficient(G, normalized=True)

# Graficar los coeficientes de Rich Club
plt.figure(figsize=(12, 6))

# Gr√°fica sin normalizar
plt.subplot(1, 2, 1)
plt.plot(list(rich_club_unnormalized.keys()), list(rich_club_unnormalized.values()), marker='o', label="Sin normalizar")
plt.title("Rich Club Coefficient (Sin normalizar)")
plt.xlabel("Grado")
plt.ylabel("Coeficiente")
plt.grid(True)
plt.legend()

# Gr√°fica normalizada
plt.subplot(1, 2, 2)
plt.plot(list(rich_club_normalized.keys()), list(rich_club_normalized.values()), marker='o', label="Normalizado")
plt.title("Rich Club Coefficient (Normalizado)")
plt.xlabel("Grado")
plt.ylabel("Coeficiente")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Identificar los nodos m√°s interconectados (hubs)
threshold = 20  # Umbral de grado (ajusta este valor seg√∫n la red)
# hub = grado arriba de media +una desviaci√≥n estandar
hubs = [node for node, degree in G.degree() if degree >= threshold]
print(f"Nodos eliminados (hubs): {hubs}")

# Crear una copia del grafo y eliminar los hubs
G_no_hubs = G.copy()
G_no_hubs.remove_nodes_from(hubs)

# Asegurarse de que no haya nodos aislados restantes
G_no_hubs.remove_nodes_from(list(nx.isolates(G_no_hubs)))

# Recalcular las m√©tricas para el grafo sin hubs
def calculate_graph_properties(graph):
    """Calcula las propiedades clave de un grafo."""
    # Coeficiente de mundo peque√±o
    try:
        avg_shortest_path = nx.average_shortest_path_length(graph)
        clustering = nx.clustering(graph)  # Coeficiente de agrupamiento
        avg_clustering = np.mean(list(clustering.values()))  # Promedio
        small_world = avg_clustering / avg_shortest_path
    except nx.NetworkXError:
        avg_shortest_path = float('inf')
        avg_clustering = 0
        small_world = 0

    # Centralidades
    closeness = nx.closeness_centrality(graph)
    betweenness = nx.betweenness_centrality(graph)

    return small_world, avg_shortest_path, avg_clustering, closeness, betweenness

# Calcular propiedades del grafo sin hubs
small_world_no_hubs, avg_shortest_path_no_hubs, avg_clustering_no_hubs, closeness_no_hubs, betweenness_no_hubs = calculate_graph_properties(G_no_hubs)

# Imprimir los resultados
print(f"Coeficiente de Mundo Peque√±o (sin hubs): {small_world_no_hubs}")
print(f"Promedio de la distancia m√°s corta (sin hubs): {avg_shortest_path_no_hubs}")
print(f"Promedio del coeficiente de agrupamiento (sin hubs): {avg_clustering_no_hubs}")

# Graficar distribuciones de centralidad en el grafo sin hubs
plt.figure(figsize=(12, 6))

# Graficar cercan√≠a
plt.subplot(1, 2, 1)
plt.hist(list(closeness_no_hubs.values()), bins=20, color='green', alpha=0.7)
plt.title("Distribuci√≥n de la Cercan√≠a (sin hubs)")
plt.xlabel("Cercan√≠a")
plt.ylabel("Frecuencia")
plt.grid(True)

# Graficar intermediaci√≥n
plt.subplot(1, 2, 2)
plt.hist(list(betweenness_no_hubs.values()), bins=20, color='blue', alpha=0.7)
plt.title("Distribuci√≥n de la Intermediaci√≥n (sin hubs)")
plt.xlabel("Intermediaci√≥n")
plt.ylabel("Frecuencia")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Si quitamos los hubs, hay una gran diferencia en la conectividad y por lo tanto la funcionalidad del cerebro se ver√≠a afectada si los quitamos, tambi√©n lo podemos ver con el √≠ndice de mundo peque√±o ya que aumentamos las distancias promedio en la red, disminuyendo la conectividad global. Se pierden la mayor√≠a de las conexiones.

```python
percentages = [10, 20, 30, 40, 50]

for p in percentages:
    G_temp = G.copy()
    num_nodes_to_remove = int(len(G) * (p / 100))

    # Calcular intermediaci√≥n y seleccionar nodos a eliminar
    betweenness = nx.betweenness_centrality(G_temp)
    nodes_to_remove = sorted(betweenness, key=betweenness.get, reverse=True)[:num_nodes_to_remove]

    # Eliminar nodos
    G_temp.remove_nodes_from(nodes_to_remove)

    # Visualizar grafo resultante
    plt.figure(figsize=(8, 6))
    nx.draw_spring(G_temp, node_size=20, edge_color='gray', alpha=0.7)
    plt.title(f"Grafo con {p}% de nodos eliminados (Mayor intermediaci√≥n)")
    plt.show()
```

# Ejercicio 10

```python
coords = mat_data['Coord']  

# Calcular la distancia geom√©trica entre los nodos
distances = squareform(pdist(coords))  # Calcula las distancias entre todas las coordenadas

# Probabilidad inversamente proporcional a la distancia
probability_matrix = 1 / (distances + np.eye(len(G.nodes)))  # Evitar divisiones por cero con np.eye
probability_matrix /= probability_matrix.sum(axis=1, keepdims=True)  # Normalizar por filas

# Crear el modelo nulo (grafo geom√©trico)
threshold = np.percentile(probability_matrix, 95)  # Ajustar el umbral al percentil deseado
connection_matrix = (probability_matrix > threshold).astype(int)  # Crear matriz de adyacencia
np.fill_diagonal(connection_matrix, 0)  # Eliminar autoconexiones

geometric_graph = nx.from_numpy_array(connection_matrix)  # Crear el grafo geom√©trico

# Funci√≥n para calcular propiedades topol√≥gicas
def calculate_topological_properties(graph):
    clustering_coeffs = nx.clustering(graph)
    degrees = dict(graph.degree())
    betweenness = nx.betweenness_centrality(graph)
    closeness = nx.closeness_centrality(graph)

    return {
        "clustering_coeffs": clustering_coeffs,
        "degrees": degrees,
        "betweenness": betweenness,
        "closeness": closeness,
    }

# Calcular propiedades topol√≥gicas del grafo geom√©trico
props_geometric = calculate_topological_properties(geometric_graph)

# Comparar con las propiedades del grafo original
props_original = calculate_topological_properties(G)

# Funci√≥n para graficar propiedades de nodos
def plot_node_properties(property_dict, title):
    plt.figure(figsize=(8, 6))
    plt.hist(list(property_dict.values()), bins=20, color='blue', alpha=0.7)
    plt.title(title)
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

# Graficar comparaci√≥n de propiedades
plot_node_properties(props_original["clustering_coeffs"], "Coeficiente de clustering (Original)")
plot_node_properties(props_geometric["clustering_coeffs"], "Coeficiente de clustering (Modelo Geom√©trico)")
plot_node_properties(props_original["betweenness"], "Centralidad de intermediaci√≥n (Original)")
plot_node_properties(props_geometric["betweenness"], "Centralidad de intermediaci√≥n (Modelo Geom√©trico)")
```

Las conexiones a larga distancia ayudan primeramente a unir dos √°reas anat√≥micamente separadas, consigo ayudan a integrar informaci√≥n, coordinar actividades y ayudar a procesos cognitivos complejos. Tambi√©n pueden servir de sustento cuando las v√≠as locales se ven afectadas adquiriendo caracter√≠sticas pl√°sticas y de resiliencia contribuyendo al mantenimiento de la eficiencia de la red a√∫n cuando haya fallos en la misma. En conclusi√≥n, las conexiones a larga distancia ayudan a un mayor funcionamiento global tanto en situaciones normales as√≠ como cuando el cerebro sufre alg√∫n da√±o.
