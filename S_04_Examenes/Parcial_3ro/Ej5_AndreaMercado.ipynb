```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import networkx as nx
import scipy.io
from matplotlib import cm 
from scipy.io import loadmat
```

```python
# Ejercicio 5: Para los datos de 'Coactivation_matrix.mat', filtre la matriz para que, para cada nodo, se mantenga aquel nodo con mayor comunicación, con el nuevo arreglo muestre los nodos y vértices del grafo

file_path = r"C:\\Users\\andim\\OneDrive\\Escritorio\\Quinto Semestre\\Modelos Computacionales\\Coactivation_matrix.mat"
mat_data = loadmat(file_path)
```

```python
mat_dict = {k: v for k, v in mat_data.items() if k[0] != '_'}
df = pd.DataFrame(mat_dict['Coactivation_matrix'])

mat_dict
```

```python
df
```

```python
G = nx.from_pandas_adjacency(df, create_using=nx.DiGraph)
```

```python
for node in G.nodes():
    edges = G[node]
    if edges:
        vecinoMax = max(edges.items(), key=lambda x: x[1]['weight'])[0]
        for vecinos in list(edges.keys()):
            if vecinos != vecinoMax:
                G.remove_edge(node, vecinos)
```

```python
nx.draw(G, node_color='purple', edge_color='orange', node_size=60)
```

```python
df_adj = nx.to_pandas_adjacency(G)
adj_matrix = df_adj.to_numpy()
```

```python
graph_list = list(G.adjacency())
graph_list
```

```python
df_adj
```

```python
x = mat_dict['Coord'][:, 0] 
y = mat_dict['Coord'][:, 1] 
z = mat_dict['Coord'][:, 2]  
```

```python
pesoMaximo = np.max(adj_matrix) if np.max(adj_matrix) > 0 else 1
pesosNormalizados = adj_matrix / pesoMaximo
colormap = cm.viridis 
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

#Plot nodes
ax.scatter3D(x, y, z, color='pink', s=60)

#Plot edges
for i in range(len(adj_matrix)):
    for j in range(len(adj_matrix)):
        weight = adj_matrix[i, j]
        if weight > 0:  #Sólo non zero edges
            color = colormap(pesosNormalizados[i, j])  #Color of edges according to weight
            ax.plot(
                [x[i], x[j]],  
                [y[i], y[j]], 
                [z[i], z[j]],  
                color=color, alpha=0.8, lw=1 + 3 * weight  
            )

#Angle
ax.view_init(elev=20., azim=30)
```
