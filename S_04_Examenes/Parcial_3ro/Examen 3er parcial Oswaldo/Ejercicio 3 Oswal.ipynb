```python
import scipy.io
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Carga del archivo
arch = scipy.io.loadmat(r"C:\Users\oswal\OneDrive\Documentos\REPO PROFE\Neurociencias-2025-1\BCT\BCT\2019_03_03_BCT\data_and_demos\fve32.mat")
G = nx.from_numpy_array(arch['CIJ'])

# Coordenadas aleatorias
x = np.random.randint(0, 10, size=32)
y = np.random.randint(0, 10, size=32)
z = np.random.randint(0, 10, size=32)

nodes = np.array([[i, j, k] for i, j, k in zip(x, y, z)])
edges = [(nodes[u], nodes[v]) for u, v in G.edges()]

# Graficar
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(111, projection="3d")

# Dibuja nodos
ax.scatter(*nodes.T, color='red', s=50)

# Dibuja aristas
for edge in edges:
    ax.plot(*zip(*edge), color='blue')

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```
