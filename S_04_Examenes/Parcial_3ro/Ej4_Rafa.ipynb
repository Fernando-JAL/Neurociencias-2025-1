```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import networkx as nx
import scipy.io
from matplotlib import cm  # Import colormap
```

```python
#Load files ig 

mat_path = r"Coactivation_matrix.mat"
mat_json = scipy.io.loadmat(mat_path)
```

```python
#File to dictionary 

mat_dict = {k: v for k, v in mat_json.items() if k[0] != '_'}
df = pd.DataFrame(mat_dict['Coactivation_matrix'])

mat_dict
```

```python
df
```

```python



binary_matadj = (mat_dict['Coactivation_matrix'] > 0.2).astype(int)
binary_matadj

```

```python
df_filtered = pd.DataFrame(binary_matadj)
df_filtered
```

```python
#Adj matrix en NetworkX
G = nx.from_pandas_adjacency(df, create_using=nx.DiGraph)
G_filtered = nx.from_pandas_adjacency(df_filtered, create_using=nx.DiGraph)
```

```python
nx.draw(G, node_color='purple', edge_color='gray', node_size=50)
```

```python
nx.draw(G_filtered, node_color='purple', edge_color='gray', node_size=50)
```

```python
#Chords!

x = mat_dict['Coord'][:, 0] 
y = mat_dict['Coord'][:, 1] 
z = mat_dict['Coord'][:, 2]  
```

```python

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')


ax.scatter3D(x, y, z, color='lightblue', s=50)


for edge in G.edges(data=True):  
    i, j = edge[0], edge[1]  
    weight = edge[2].get('weight', 0)  

    if weight > 0:  
        ax.plot(
            [x[i], x[j]], 
            [y[i], y[j]],  
            [z[i], z[j]],  
            color='gray', alpha=0.5 + 0.5 * weight, lw=1 + 2 * weight  
        )


ax.view_init(elev=20., azim=30)

plt.show()
```
