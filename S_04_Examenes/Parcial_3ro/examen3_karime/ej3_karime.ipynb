```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io
import networkx as nx
from matplotlib import animation
```

```python
math_path = r"C:\Users\karim\OneDrive\Documentos\Python Scripts\fve30.mat"
contents = scipy.io.loadmat(math_path)
G = nx.from_numpy_array(contents['CIJ'])

x = np.random.randint(0, 51, size= 32)
y = np.random.randint(0, 51, size= 32)
z = np.random.randint(0, 51, size= 32)

nodes = np.array([[i, j, k] for i, j, k in zip(x, y, z)])
edges = np.array([(nodes[u], nodes[v]) for u, v in G.edges()])

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection="3d")

def init():
    ax.scatter(*nodes.T) 
    for edge in edges:
        ax.plot(*edge.T)

    return fig,

def animate(i):
    ax.view_init(elev=20, azim=i*4)
    return fig,

animation = animation.FuncAnimation(fig, animate, init_func=init, frames=90, interval=200, blit=False)
plt.show()
```
