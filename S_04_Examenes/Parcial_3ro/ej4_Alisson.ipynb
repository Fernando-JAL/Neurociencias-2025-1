### Ejercicio 4 
Para los datos de 'Coactivation_matrix.mat', filtre la matriz para obtener los valores > 0.2, con el nuevo arreglo muestre los nodos y vértices del grafo 

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
from scipy.io import loadmat
from mpl_toolkits.mplot3d import Axes3D  # Necesario para 3D

# Cargar la matriz de coactivación
data = loadmat(r'C:\Users\aliss\Documents\Repo_modelos\pythonProject\TAREAS\Coactivation_matrix.mat')
adj_matrix = data['Coactivation_matrix']

# Filtrar valores menores o iguales a 0.2
adj_matrix[adj_matrix <= 0.2] = 0

# Crear el grafo a partir de la matriz de adyacencia
G = nx.from_numpy_array(adj_matrix)

# Generar coordenadas aleatorias para los nodos en 3D
num_nodes = G.number_of_nodes()
x = np.random.randint(0, 50, size=num_nodes)
y = np.random.randint(0, 50, size=num_nodes)
z = np.random.randint(0, 50, size=num_nodes)

# Convertir las posiciones a un array
nodes = np.array([x, y, z]).T

# Obtener las aristas en formato 3D
edges = [(nodes[u], nodes[v]) for u, v in G.edges()]

# Crear la figura y el eje 3D
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection="3d")

# Graficar nodos
ax.scatter(nodes[:, 0], nodes[:, 1], nodes[:, 2], c='blue', s=100, label='Nodos')

# Graficar aristas
for edge in edges:
    x_coords = [edge[0][0], edge[1][0]]
    y_coords = [edge[0][1], edge[1][1]]
    z_coords = [edge[0][2], edge[1][2]]
    ax.plot(x_coords, y_coords, z_coords, c='gray')

# Configurar la vista y etiquetas
ax.set_xlim([0, 50])
ax.set_ylim([0, 50])
ax.set_zlim([0, 50])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.legend()

# Mostrar el gráfico
plt.show()
```
