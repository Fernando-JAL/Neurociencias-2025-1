Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str


```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import seaborn as sns


def df_ordered (dataframe, column_name):
    return dataframe.sort_values(by=column_name, ascending=False)

```

Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

```python

  def plot_bar(dataframe, column_name):
    dataframe[column_name].value_counts().plot(kind='bar')
    plt.title(f'Gráfico de barras de {column_name}')
    plt.xlabel(column_name)
    plt.ylabel('Frecuencia')
    plt.show()
```

Países con mayor depresión

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt



mip_path = "/Users/marianazuniga/Downloads/1- mental-illnesses-prevalence.csv"
df_mip = pd.read_csv(mip_path, sep = ",")



# Filtro de las columnas
depression_df= df_mip[['Entity', 'Year', 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized']]
depression_df.columns = ['País', 'Año', 'Porcentaje de Depresion']
df_recent = depression_df[depression_df['Año'] == depression_df['Año'].max()]
df_recent = df_recent.dropna(subset=['Porcentaje de Depresion'])


#Orden
df_orden = df_recent.sort_values(by='Porcentaje de Depresion', ascending=False).head(10)


#Resultados
print(df_orden)



```

Gráfica de barras mostrando el valor de depresión y la entidad (país)

```python

plt.figure(figsize=(10, 6))
plt.barh(df_orden['País'], df_orden['Porcentaje de Depresion'], color= 'orange')


plt.gca().invert_yaxis()


plt.title('Top 10 Países con Mayor Prevalencia de Desórdenes Depresivos (Año más reciente)')
plt.xlabel('Porcentaje de la población afectada (%)')
plt.ylabel('País')


plt.tight_layout()
plt.show()
```

```python
csv_path = "/Users/marianazuniga/Downloads/cancer_reg.csv"

df_cancer = pd.read_csv(csv_path, sep = ",")
```

La correlación entre las features [ 'target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate', 'medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage' ]

```python

features = ['target_deathrate', 'avganncount', 'avgdeathsperyear', 
            'incidencerate', 'medincome', 'povertypercent', 
            'pctprivatecoverage', 'pctpubliccoverage']


correlation_matrix = df_cancer[features].corr()

print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlación')
plt.show()
```

Mostrar la distribución de cada feature del dataframe anterior

```python
plt.figure(figsize=(16, 10))

# Crear un gráfico para cada característica
for i, feature in enumerate(features):
    plt.subplot(3, 3, i + 1)  # Cambia los números para ajustar el diseño
    plt.hist(df_cancer[feature], bins=30, color='orange', alpha=0.7)  # Histograma
    plt.title(f'Distribución de {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')

# Ajustar el diseño
plt.tight_layout()
plt.show()
```

Aquellos que dependencia lineal encontrar los valores de la recta. hint: scipy

```python
features = ['avganncount', 'avgdeathsperyear', 'incidencerate', 
            'medincome', 'povertypercent', 'pctprivatecoverage', 
            'pctpubliccoverage']


# Variable dependiente
y = df_cancer['target_deathrate']

# Configurar el tamaño de la figura
n = len(features)
fig, axes = plt.subplots(nrows=2, ncols=( n+1) // 2, figsize=(20, 15))

# Aplanar el array de ejes para facilitar la iteración
axes = axes.flatten()

# Iterar sobre cada característica para calcular la regresión lineal
for i, feature in enumerate(features):
    x = df_cancer[feature]

    # Calcular la regresión lineal
    slope, intercept, r_value, p_value, std_err = linregress(x, y)

    # Graficar los datos y la línea de regresión
    axes[i].scatter(x, y, color='blue', label='Datos')
    axes[i].plot(x, slope * x + intercept, color='red', label='Línea de regresión')
    axes[i].set_title(f'Regresión Lineal: {feature} vs Tasa de Mortalidad')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Tasa de Mortalidad')
    axes[i].legend()
    axes[i].grid()

# Ajustar el diseño
plt.tight_layout()
plt.show()
```
