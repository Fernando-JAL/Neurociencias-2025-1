```python
# Cargamos la base de datos
import pandas as pd
ruta_archivo = r"C:\Users\ANEL\OneDrive\Escritorio\database\4- adult-population-covered-in-primary-data-on-the-prevalence-of-mental-illnesses.csv"
df = pd.read_csv(ruta_archivo)

# Cargamos para ver como se ven las primeras filas
df.head(11)
```

```python
# 1) Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor 
#considerando la característica str.

# Definir la función df_ordered
def df_ordered(dataframe, nombre_columna):
    if nombre_columna in dataframe.columns:
        return dataframe.sort_values(by=nombre_columna, ascending=False)
    else:
        return f"La columna '{nombre_columna}' no existe en el dataframe."
```

```python
# Probar la función con el dataframe y la columna Eating disorders
df_ordered_result = df_ordered(df, 'Eating disorders')
df_ordered_result.head(11)
## eeeee si sale ordenada !!!!
```

```python
import matplotlib.pyplot as plt

# Definir la función plot_bar
def plot_bar(dataframe, nombre_columna):
    if nombre_columna in dataframe.columns:
        dataframe[nombre_columna] = pd.to_numeric(dataframe[nombre_columna], errors='coerce')
        
        # Ordenamos de mayor a menor 
        dataframe_ordenado = dataframe.sort_values(by=nombre_columna, ascending=False)
        
        # Graficar las barras
        plt.figure(figsize=(10, 6))
        plt.bar(dataframe_ordenado['Entity'], dataframe_ordenado[nombre_columna], color='olivedrab')
        plt.xlabel('Entity')
        plt.ylabel(nombre_columna)
        plt.title(f'Gráfico de barras de: {nombre_columna}')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f"La columna '{nombre_columna}' no existe en el dataframe.")
```

```python
# probamos con 'Bipolar disorder'
plot_bar(df, 'Bipolar disorder')
```

```python
# Países con mayor DEPRESIÓN: gráfica de barras mostrando el valor de depresión y la entidad
plot_bar(df, 'Major depression')
```

```python
# Países con mayor DESORDEN ALIMENTICIO: gráfica de barras mostrando el valor de desorden alimenticio y la entidad (país)
plot_bar(df, 'Eating disorders')
```

```python
# países con mayor ESQUIZOFRENIA: gráfica de barras mostrando el valor de esquizofrenia y la entidad (país):
plot_bar(df, 'Schizophrenia')
```

```python
# crear un data frame con los valores por país de
# ["Entity Code", "Year", "Schizophrenia disorders", "Depressive disorders", "Anxiety disorders", "Bipolar disorders", "Eating disorders"]

columnas = ["Entity", "Code", "Year", "Schizophrenia", "Major depression", "Anxiety disorders", "Bipolar disorder", "Eating disorders"]
df_selected = df[columnas]

# Mostrar los primeros registros del nuevo dataframe
df_selected.head(22)
```

```python
# mostrar la distribución de cada feature del dataframe de arriba

# Crear gráficos de distribución para cada columna numérica
import seaborn as sns
plt.figure(figsize=(15, 10)) # Establecer el tamaño de la figura

# Lista de las columnas a visualizar
columns_to_plot = ["Schizophrenia", "Major depression", "Anxiety disorders", "Bipolar disorder", "Eating disorders"]

columns_sorted = df_selected[columns_to_plot].mean().sort_values(ascending=False).index

# Crear subplots para cada característica
for m, col in enumerate(columns_sorted, 1):
    plt.subplot(2, 3, m)
    sns.histplot(df_selected[col], kde=True, color='olive')
    plt.title(f'Distribución de: {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# mostrar en un mapa de color la correlación entre las features:
# ["Schizophrenia disorders", "Depressive disorders", "Anxiety disorders", "Bipolar disorders", "Eating disorders"]

import seaborn as sns
import matplotlib.pyplot as plt

# Selecting relevant columns for correlation analysis
col_data = ["Major depression", "Bipolar disorder", "Eating disorders", "Schizophrenia", "Anxiety disorders"]

# Cleaning the data by dropping rows with missing values
df_nueva = df[col_data].dropna()

# Calculating the correlation matrix
corr_mat = df_nueva.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_mat, annot=True, cmap='Greens', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Mental Health Disorders')
plt.show()
```

```python
#Del dataset cancer_reg determinar:
# la correlación entre las features:
# [ 'target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate', 'medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage' ]


# primero cargamos los datos 

import pandas as pd
archivo2 = r"C:\Users\ANEL\OneDrive\Escritorio\database\cancer_reg.csv"
df = pd.read_csv(archivo2)

# Cargamos para ver como se ven las primeras filas
df.head()
```

```python
# la correlación entre las features:
# [ 'target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate', 'medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage' ]

# Selecting the relevant columns for correlation
columnas_sel = ['target_deathrate', 'avganncount', 'avgdeathsperyear','incidencerate', 'medincome', 'povertypercent','pctprivatecoverage', 'pctpubliccoverage']

# Calculating the correlation matrix
matrix = df[columnas_sel].corr()

matrix
```

```python
# mostrar la distribución de cada feature del dataframe anterior
import matplotlib.pyplot as plt
import seaborn as sns

# hacemos el plot
plt.figure(figsize=(15, 10))
# hacer la distribución
for i, column in enumerate(columnas_sel, 1):
    plt.subplot(4, 2, i)
    sns.histplot(df[column], kde=True, bins=30, color= 'orangered')
    plt.title(f'Distribución de: {column}')
    plt.tight_layout()

# Show the plots
plt.show()
```

```python
# aquellos que dependencia lineal encontrar los valores de la recta. hint: scipy

from scipy.stats import linregress

# Definir un umbral para considerar una relación lineal fuerte
threshold = 0.7 

# Encontrar las parejas de variables con correlación fuerte
corr_alta = [(col1, col2) for col1 in columnas_sel for col2 in columnas_sel 
                   if col1 != col2 and abs(matrix.loc[col1, col2]) > threshold]

# Calcular la recta de regresión para cada par de variables altamente correlacionadas
for col1, col2 in corr_alta:
    slope, intercept, r_value, p_value, std_err = linregress(df[col1], df[col2])
    print(f"Linear regression between {col1} and {col2}:")
    print(f"Slope: {slope}, Intercept: {intercept}, R-squared: {r_value**2}\n")
```
