1. Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str


```python
def df_ordered(df, str):
    df_sorted = df.sort_values(by= str, ascending = False)
    return df_sorted
```

2. Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

```python
def plot_bar (df, str):
    df.plot(kind='bar', y=str, x='Entity', legend=False, color='hotpink')
    plt.ylabel('Prevalence')
    plt.xlabel('Entity')
    plt.title(str)
    plt.show()
```

```python
threshold = 0.8
strong_corr = corr[(corr > threshold) | (corr < -threshold)]
strong_corr
```

```python
from scipy.stats import linregress

w = df_filter['avgdeathsperyear']
x = df_filter['avganncount']
y = df_filter['pctprivatecoverage']
z = df_filter['povertypercent']
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x, w)
print(f"Pendiente (slope): {slope}")
print(f"Intersección (intercept): {intercept}")
print(f"Coeficiente de correlación (r_value): {r_value}")
print(f"P-valor (p_value): {p_value}")
print(f"Error estándar (std_err): {std_err}")
```

```python
slope, intercept, r_value, p_value, std_err = linregress(x, w)
plt.scatter(x, w)
plt.plot(x, intercept + slope * x, color='red')
plt.xlabel('avganncount')
plt.ylabel('avgdeathsperyear')
plt.title('Regresión Lineal entre avganncount y avgdeathsperyear')
plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = linregress(y, z)
print(f"Pendiente (slope): {slope}")
print(f"Intersección (intercept): {intercept}")
print(f"Coeficiente de correlación (r_value): {r_value}")
print(f"P-valor (p_value): {p_value}")
print(f"Error estándar (std_err): {std_err}")
```
