- Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str

```python
def df_ordered(df: pd.DataFrame, feature: str):
    df.sort_values(by=feature, ascending=False, inplace=True)
```

- Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

```python
def plot_bar(df: pd.DataFrame, feature1: str, feature2: str):
    #sns.barplot(data=df, y=feature1, x=feature2, orient='h')
    df.plot.barh(x=feature1, y=feature2)
```

```python
df_filtered = df_cancer[[ 'target_deathrate', 'avganncount', 'avgdeathsperyear', 
                         'incidencerate', 'medincome', 'povertypercent', 
                         'pctprivatecoverage', 'pctpubliccoverage' ]]
df_filtered.head(3)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for idx, column in enumerate(df_filtered):
    plt.subplot(3, 3, idx+1)
    sns.histplot(data=df_filtered, x=column, kde=True)
    plt.title('histogram')
    
ax = plt.subplot(3, 3, 9)
ax.axis('off')
```

```python
sns.pairplot(df_filtered, markers='o')
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df_filtered.avgdeathsperyear,
                                                               df_filtered.avganncount)
```

```python
slope, intercept, r_value, p_value, std_err
```

```python
fig = plt.figure(figsize=(10, 5))
#df_cancer.plot.scatter(y='avganncount', x='avgdeathsperyear')
sns.scatterplot(data=df_cancer, y='avganncount', x='avgdeathsperyear')

x = range(0, 14000)
y = slope*x + intercept
plt.plot(x, y, 'm')
```

```python
fig = plt.figure(figsize=(15, 10))
sns.lmplot(x="avgdeathsperyear", y="avganncount", data=df_filtered)

x = range(0, 14000)
y = slope*x + intercept
plt.plot(x, y, 'm--')
plt.show()
```
