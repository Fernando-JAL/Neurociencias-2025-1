```python
# Crear función "df_ordered" que reciba un dataframe y str y haga el plot de dataframe ordenado de mayor a menor considerando la característica str
def df_ordered(dataframe, nombre_columna):
    if nombre_columna in dataframe.columns:
        return dataframe.sort_values(by=nombre_columna, ascending=False)
    else:
        return f"La columna '{nombre_columna}' no existe en el dataframe"
```

```python
# Probar la función con el dataframe y la columna Bipolar disorder
df_ordered_result = df_ordered(df, 'Bipolar disorder')
df_ordered_result.head(11)
```

```python
import matplotlib.pyplot as plt

# Definir la función plot_bar
def plot_bar(dataframe, nombre_columna):
    if nombre_columna in dataframe.columns:
        dataframe[nombre_columna] = pd.to_numeric(dataframe[nombre_columna], errors= 'coerce')

        #orden
        dataframe_ordenado = dataframe.sort_values(by=nombre_columna, ascending=False)
        
        plt.figure(figsize=(10, 6))
        plt.bar(dataframe_ordenado['Entity'], dataframe_ordenado[nombre_columna], color='olivedrab')
        plt.xlabel('Entity')
        plt.ylabel(nombre_columna)
        plt.title(f' Grafico de barras de: {nombre_columna}')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f"La columna '{nombre_columna}' no existe en el dataframe.")
```

```python
# prueba con Bipolar disorder
plot_bar(df, 'Bipolar disorder')
```

```python
# Países con mayor DEPRESIÓN: gráfica de barras mostrando el valor de depresión y la entidad
plot_bar(df, 'Major depression')
```

```python
# Países con mayor DESORDEN ALIMENTICIO: gráfica de barras mostrando el valor de desorden alimenticio y la entidad (país)
plot_bar(df, 'Eating disorders')
```

```python
# países con mayor esquizofrenia: gráfica de barras mostrando el valor de esquizofrenia y la entidad (país):
plot_bar(df, 'Schizophrenia')
```

```python
# la correlación entre las features:
# [ 'target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate', 'medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage' ]

# Selecting the relevant columns for correlation
columnas_sel = ['target_deathrate', 'avganncount', 'avgdeathsperyear','incidencerate', 'medincome', 'povertypercent','pctprivatecoverage', 'pctpubliccoverage']

# Calculating the correlation matrix
matrix = df[columnas_sel].corr()

matrix
```

```python
##%%
# mostrar la distribución de cada feature del dataframe anterior
import matplotlib.pyplot as plt
import seaborn as sns

# hacer plot
plt.figure(figsize=(10, 10))

# hacer distribución
for i, column in enumerate(columnas_sel, 1):
    plt.subplot(4, 2, i)
    sns.histplot(df[column], kde=True, bins=30, color='#FFB2D8')
    plt.title(f'Distribución de: {column}')
    plt.tight_layout()

# Show the plots
plt.show()
```

```python
# aquellos que dependencia lineal encontrar los valores de la recta. hint: scipy

from scipy.stats import linregress

# Definir un umbral para considerar una relación lineal fuerte
threshold = 0.7 

# Encontrar las parejas de variables con correlación fuerte
corr_alta = [(col1, col2) for col1 in columnas_sel for col2 in columnas_sel 
                   if col1 != col2 and abs(matrix.loc[col1, col2]) > threshold]

# Calcular la recta de regresión para cada par de variables altamente correlacionadas
for col1, col2 in corr_alta:
    slope, intercept, r_value, p_value, std_err = linregress(df[col1], df[col2])
    print(f"Linear regression between {col1} and {col2}:")
    print(f"Slope: {slope}, Intercept: {intercept}, R-squared: {r_value**2}\n")
```
