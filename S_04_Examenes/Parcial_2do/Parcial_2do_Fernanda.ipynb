# Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str

# Países con mayor depresión

```python
import pandas as pd
```

```python
df = pd.read_csv('4- adult-population-covered-in-primary-data-on-the-prevalence-of-mental-illnesses (1).csv')
```

```python
print(df.columns)
```

```python
def df_ordered(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
  return df.sort_values(by=column_name, ascending=False)
```

```python
df_ordenado = df_ordered(df, 'Major depression')
df_filtrado = df_ordenado[['Entity', 'Major depression']]
print(df_filtrado)
```

# Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

# Gráfica de barras mostrando el valor de depresión y la entidad (país)

```python
import matplotlib.pyplot as plt
```

```python
def plot_bar(df: pd.DataFrame, Major_depression: str, Entity: str) -> None:
  df_sorted = df.sort_values(by=Major_depression, ascending=False)

  values = df_sorted[Major_depression]
  labels = df_sorted[Entity]

  plt.figure(figsize=(8,4))
  plt.bar(labels, values)

  plt.title(f'Gráfico de barras de {Major_depression} por {Entity}')
  plt.xlabel('Major depresión')
  plt.ylabel('País')

  plt.xticks(rotation=90)
  plt.tight_layout()
  plt.show()
plot_bar(df_ordenado, 'Major depression', 'Entity')
```

# Mostrar en un mapa de color la correlación entre las features ["Schizophrenia disorders", "Depressive disorders", "Anxiety disorders", "Bipolar disorders", "Eating disorders"]

# Aquellos con dependencia lineal, encontrar los valores de la recta. hint: scipy

```python
import pandas as pd
import scipy.stats as spicy
import matplotlib.pyplot as plt

def linear_regression(feature1, feature2, df):
  x = df[feature1]
  y = df[feature2]

  slope, intercept, r_value, p_value, std_err = spicy.linregress(x, y)

  print(f"Regresión lineal entre {feature1} y {feature2}:")
  print(f"Pendiente (slope): {slope}")
  print(f"Intersección (intercept): {intercept}")
  print(f"Coeficiente de correlación (r-value): {r_value}")
  print(f"Valor p (p-value): {p_value}")
  print(f"Error estándar de la pendiente: {std_err}")

  plt.figure(figsize=(8,6))
  plt.scatter(x, y, label=f'{feature1} vs {feature2}', color='blue')
  plt.plot(x, slope * x + intercept, color='red', label=f'Regresión lineal')
  plt.title(f'Relación lineal entre {feature1} y {feature2}')
  plt.xlabel(feature1)
  plt.ylabel(feature2)
  plt.legend()
  plt.show()


df = pd.read_csv('cancer_reg.csv')

df_filtrado = df[['target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate',
                  'medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage']]

# Calcular la matriz de correlación de Pearson
correlation_matrix = df_filtrado.corr(method='pearson')

#filtered_correlation = correlation_matrix[(correlation_matrix > 0.6) | (correlation_matrix < -0.6)]
# Mostrar los coeficientes filtrados
#print(filtered_correlation)


linDep1 = ("avganncount", "avgdeathsperyear")
linDep2 = ("povertypercent", "medincome")
linDep3 = ("pctprivatecoverage", "medincome")
linDep4 = ("pctpubliccoverage", "medincome")
linDep5 = ("pctprivatecoverage", "povertypercent")
linDep6 = ("pctpubliccoverage", "povertypercent")
linDep7 = ("pctpubliccoverage", "pctprivatecoverage")

linDeps = [linDep1, linDep2, linDep3, linDep4, linDep5, linDep6, linDep7]
for lindep in linDeps:
  linear_regression(lindep[0], lindep[1], df_filtrado)
```
