## 1. Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str

```python
def df_ordered(df, str):  
    # Ordenar el DataFrame de mayor a menor
    return df.sort_values(by=str, ascending=False) # False pa que vaya de mayor a menor
```

## 2. Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

```python
def plot_bar(df, str):
    df.plot(kind='bar', x='Entity', y=str)

    # Título y etiquetas
    plt.title(f'Gráfico de Barras de {str}')
    plt.xlabel('Entity')
    plt.ylabel(str)
    plt.xticks(rotation=90)  # Rotar etiquetas del eje x si es necesario
    plt.show()
```

## 2. Mostrar la distribución de cada feature del dataframe anterior

```python
from scipy.stats import linregress
cols_high = ['target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate','medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage']
df_filter = df_filter[cols_high]
corr = df_filter.corr()
pairplot=sns.pairplot(df_filter, markers='o')

# Valores de la recta de los features con dependencia lineal 
x= df_filter["avganncount"]
y= df_filter["avgdeathsperyear"]

slope, intercept, r_value, p_value, std_err = linregress(x, y)
print(f"Pendiente (slope): {slope}")
print(f"Intersección (intercept): {intercept}")
print(f"Coeficiente de correlación (r_value): {r_value}")
print(f"P-valor (p_value): {p_value}")
print(f"Error estándar (std_err): {std_err}")
```

## 3. Aquellos que dependencia lineal encontrar los valores de la recta. hint: scipy

```python
slope, intercept, r_value, p_value, std_err = linregress(x, y)
plt.scatter(x, y)
plt.plot(x, intercept + slope * x, color='red')
plt.xlabel('avganncount')
plt.ylabel('avgdeathsperyear')
plt.title('Regresión lineal entre avganncount y avgdeathsperyear')
plt.show()
```
