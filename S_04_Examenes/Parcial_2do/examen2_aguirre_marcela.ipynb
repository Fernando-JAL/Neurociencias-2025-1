```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv_path = r"4- adult-population-covered-in-primary-data-on-the-prevalence-of-mental-illnesses.csv"
mental_illness = pd.read_csv(csv_path)
mental_illness.loc[17, 'Schizophrenia']=0.1
mental_illness['Schizophrenia'] = pd.to_numeric(mental_illness['Schizophrenia'])
mental_illness.describe()
```

```python
def df_ordered (df, str):
    return df.sort_values(by = str, ascending=False)

# adapté este codigo para poder utilizarlo para evaluar la columna de interés según los países 'Entity'
def plot_bar (df, str):
    df.plot(kind='bar', y=str, x='Entity', legend=False, color='magenta')
    plt.ylabel('Prevalence')
    plt.xlabel('Entity')
    plt.title(str)
```

```python
depression = df_ordered(mental_illness, 'Major depression')
depression
```

```python
plot_bar(depression, 'Major depression')
```

```python
ed = df_ordered(mental_illness, 'Eating disorders')
ed
```

```python
plot_bar(ed, 'Eating disorders')
```

```python
#mental_illness.loc[mental_illness['Entity'] == 'Southern Sub-Saharan Africa', 'Schizophrenia'] = 0.1
esquizofrenia = df_ordered(mental_illness, 'Schizophrenia')
esquizofrenia
```

```python
plot_bar(esquizofrenia, 'Schizophrenia')
```

```python
csv_path=r"1- mental-illnesses-prevalence.csv"
DF = pd.read_csv(csv_path)
DF=DF.drop('Entity', axis=1, inplace=False)
DF=DF.rename(columns={'Code':'Entity code', 'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized':'Schizophrenia disorders', 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized':'Depressive disorders', 'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized':'Anxiety disorders', 'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized':'Bipolar disorders', 'Eating disorders (share of population) - Sex: Both - Age: Age-standardized':'Eating disorders'})
DF.describe()
```

```python
sns.histplot(DF['Depressive disorders'], stat='count', kde=True, color='orangered')
plt.xlabel('Depressive disorders Index')
plt.ylabel('Frecuencia')
plt.title('Distribución índice de depresión')
```

```python
sns.histplot(DF['Schizophrenia disorders'], stat='count', kde=True, color='magenta')
plt.xlabel('Schizophrenia disorders index')
plt.ylabel('Frecuencia')
plt.title('Distribución índice de esquizofrenia')
```

```python
sns.histplot(DF['Anxiety disorders'], stat='count', kde=True, color = 'darkviolet')
plt.xlabel('Anxiety disorders index')
plt.ylabel('Frecuencia')
plt.title('Distribución índice de ansiedad')
```

```python
sns.histplot(DF['Bipolar disorders'], stat='count', kde=True, color = 'yellow')
plt.xlabel('Bipolar disorders index')
plt.ylabel('Frecuencia')
plt.title('Distribución índice de bipolaridad')
```

```python
sns.histplot(DF['Eating disorders'], stat='count', kde=True, color='violet')
plt.xlabel('Eating disorders index')
plt.ylabel('Frecuencia')
plt.title('Distribución índice de trastornos alimenticios')
```

```python
# Parte 2
csv_path = r"cancer_reg.csv"
cancer = pd.read_csv(csv_path, sep=",")
filter = ['target_deathrate', 'avganncount', 'avgdeathsperyear', 'incidencerate', 'medincome', 'povertypercent', 'pctprivatecoverage', 'pctpubliccoverage']
df_filter=cancer[filter]
corr=df_filter.corr()
corr.style.background_gradient(cmap='spring')
# elegiré las correlaciones mayores a 0.6 y menores a -0.6
```

```python
plt.figure(figsize=(16, 8))
for i in range(8):
    plt.subplot(2,4,i+1)
    sns.histplot(df_filter.iloc[:,i], stat='count', kde=True, color='magenta')
plt.tight_layout()  # Ajustar los subplots
plt.show()
```

```python
sns.pairplot(df_filter,markers='o', plot_kws={'color': 'darkviolet'})
```

```python
slope, intercept, r_value, p_value, std_err= stats.linregress(x=df_filter['avganncount'], y=df_filter['avgdeathsperyear'])
print('Dependencia lineal avganncount y avgdeathsperyear:','Pendiente',slope,'Ordenada al origen', intercept)
```
