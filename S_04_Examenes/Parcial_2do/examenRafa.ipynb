```python
#Filtrado solo valores 2019 para dataframe de csv 1
df1_filtrado2019 = df1[df1['Year'] > 2018]

df1_filtrado2019
```

```python
#Función de sorteado

def df_ordered(dataframe, nombre_columna):
    
    if nombre_columna in dataframe.columns:
        return dataframe.sort_values(by=nombre_columna, ascending=False)
    
    else:
        return print("No existe esa columna")
```

```python
#Función de bar 

import matplotlib.pyplot as plt

def plot_bar(dataframe, columna):
    
    if columna in dataframe.columns:
        plt.figure(figsize=(10, 6))
        plt.bar(dataframe['Entity'], dataframe[columna], color='#696ac9')
        plt.xlabel('Country')
        plt.ylabel(columna)
        plt.title(f': {columna}')
        plt.xticks(rotation=45, fontsize=2)

        plt.show()
        
    else:
        
        print(f"La columna '{nombre_columna}' no existe en el dataframe.")
```

```python
df_depression = df_ordered(df1_filtrado2019 , "Depressive disorders (share of population) - Sex: Both - Age: Age-standardized")

val_depression = ["Entity","Depressive disorders (share of population) - Sex: Both - Age: Age-standardized"]

df_depression_alt = df_depression[val_depression]
df_depression_alt.reset_index(drop=True, inplace=True)
df_depression_alt
```

```python
plot_bar(df_depression, 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized')
```

```python
df_eatingdisorders = df_ordered(df1_filtrado2019 , 'Eating disorders (share of population) - Sex: Both - Age: Age-standardized')

val_eatingdisorders = ["Eating disorders (share of population) - Sex: Both - Age: Age-standardized"]

df_eatingdisorders_alt = df_eatingdisorders[val_eatingdisorders]
df_eatingdisorders_alt.reset_index(drop=True, inplace=True)
df_eatingdisorders_alt
```

```python
plot_bar(df_eatingdisorders, 'Eating disorders (share of population) - Sex: Both - Age: Age-standardized')
```

```python
df_schizophrenia = df_ordered(df1_filtrado2019 , 'Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized')

val_schizophrenia = ["Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized"]

df_schizophrenia_alt = df_schizophrenia[val_schizophrenia]
df_schizophrenia_alt.reset_index(drop=True, inplace=True)
df_schizophrenia_alt
```
