```python
# Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str

def df_ordered(df, column_name):
    return df.sort_values(by=column_name, ascending=False)

print(adult1_data.columns)
df_ordenado = df_ordered(adult1_data, "Major depression")
print(df_ordenado)


```

```python
# Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

def plot_bar(df_name, x_column, y_column, color):
    df = globals().get(df_name)
    
    if df is None or x_column not in df.columns or y_column not in df.columns:
        raise ValueError("DataFrame o columnas inválidas.")
    
    x_data = df[x_column]
    y_data = df[y_column]
    
    plt.figure(figsize=(10, 6))
    plt.bar(x_data, y_data, color=color)
    plt.title(f'{y_column} vs {x_column}', fontsize=14)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```
