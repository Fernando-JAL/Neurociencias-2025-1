3. Aquellos con dependencia lineal, encontrar los valores de la recta.hint: scipy

```python
from scipy.stats import linregress
dep_features = ['avganncount', 'avgdeathsperyear']

ft_dependientes = df_cancer[dep_features]
corr = ft_dependientes.corr()
corr

df= pd.DataFrame(ft_dependientes)
df.to_csv("dependientes_lineales")

df = pd.read_csv("dependientes_lineales")

x = df["avganncount"]
y = df["avgdeathsperyear"]

slope, intercept, r_value, p_value, std_err = linregress(x, y)

print(f"Pendiente (slope): {slope}")
print(f"Intersección (intercept): {intercept}")
print(f"Coeficiente de correlación (r_value): {r_value}")
print(f"Error estándar (std_err): {std_err}")

y_pred = slope * x + intercept

plt.scatter(x, y, label='Datos')
plt.plot(x, y_pred, color='red', label=f'Recta de mejor ajuste: y = {slope:.2f}x + {intercept:.2f}')
plt.title("Dependencia lineal")
plt.legend()
plt.grid(True)
plt.show()
```
