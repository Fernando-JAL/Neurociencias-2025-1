# Exploratory Data Analysis

```python
plt.figure(figsize=(20, 10))
df_cancer.hist(column="target_deathrate", bins=30)

plt.show()
```

```python
plt.figure(figsize=(20, 10))

df_cancer.plot.scatter(x='medincome', y='target_deathrate')
plt.xlabel('medincome')
plt.ylabel('target_deathrate')
plt.show()
```

```python
df_cancer.head()
```

```python
plt.figure(figsize=(20, 10))

df_cancer.hist(column='povertypercent', bins=30)
```

```python
top10_poverty_states = df_cancer.sort_values(by='povertypercent', ascending=False).head(10)
top10_poverty_states
```

```python
plt.figure(figsize=(20, 10))

top10_poverty_states.plot.bar(x='geography', y='povertypercent', rot=45)
plt.ylabel('Percentage [%]')
plt.show()
```

```python
plt.figure(figsize=(10, 5))

corr = df_cancer.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
cols_low = ['target_deathrate', 'incidencerate', 'pctunemployed16_over', 'pctpubliccoverage', 'povertypercent']
cols_high = ['avganncount', 'avgdeathsperyear', 'target_deathrate', 'incidencerate', 'medincome']
```

```python
df_cancer_low = df_cancer[cols_low]
corr = df_cancer_low.corr()
corr
```

```python
import seaborn as sns
```

```python
sns.pairplot(df_cancer_low, markers='o')
```

```python
df_cancer_high = df_cancer[cols_high]
corr = df_cancer_high.corr()
corr
```

```python
sns.pairplot(df_cancer_high, markers='o')
```
