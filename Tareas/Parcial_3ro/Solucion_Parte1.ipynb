```python
# Se define la semilla
np.random.seed(seed=5002)
```

```python
# Ej1. Generar un número aleatorio entero en el rango [-3, 3]
mu = np.random.randint(-3, 3)
# Ej2. Generar un número aleatorio flotante en el rango [-10, 10]
sigma = np.random.uniform(-10, 10)
print(mu, sigma)
# Ej3. Genera un array de 1000 números con distribución normal
data = np.random.normal(loc=mu, scale=sigma, size=1000)
```

```python
# Ej 4. Plotear la distribución
bin_edges = np.histogram_bin_edges(data, bins='scott')
sns.histplot(data, bins=bin_edges, kde=True)
plt.vlines(x=[mu-sigma, mu+sigma], ymin=0, ymax=120, color='green')
plt.vlines(x=[mu-2*sigma, mu+2*sigma], ymin=0, ymax=120, color='yellow')
plt.vlines(x=[mu-3*sigma, mu+3*sigma], ymin=0, ymax=120, color='red')
```

```python
# Ej 5. Calcular sigmas
sigma1 = np.std(data)  # sigma1 = sigma
sigma2 = 2*sigma1
sigma3 = 3*sigma1
print(sigma1, sigma2, sigma3)
```

```python
def tarea(mu, sigma, array_length):
    data = np.random.normal(mu, sigma, size=array_length)
    
    # Ej 4. Plotear la distribución
    bin_edges = np.histogram_bin_edges(data, bins='scott')
    sns.histplot(data, bins=bin_edges, kde=True)
    plt.vlines(x=mu, ymin=0, ymax=120, color='black')
    plt.vlines(x=[mu-sigma, mu+sigma], ymin=0, ymax=120, color='green')
    plt.vlines(x=[mu-2*sigma, mu+2*sigma], ymin=0, ymax=120, color='yellow')
    plt.vlines(x=[mu-3*sigma, mu+3*sigma], ymin=0, ymax=120, color='red')
    
    # Ej 5. Calculo de sigmas
    sigma1 = np.std(data)  # sigma1 = sigma
    sigma2 = 2*sigma1
    sigma3 = 3*sigma1
    
    # Ej 6-8. Calcular el porcentaje de números en el rango de sigmas
    porcentaje_1sigma = sum(((mu-sigma1) < data) & (data < (mu+sigma1)))/len(data)
    porcentaje_2sigma = sum(((mu-sigma2) < data) & (data < (mu+sigma2)))/len(data)
    porcentaje_3sigma = sum(((mu-sigma3) < data) & (data < (mu+sigma3)))/len(data)
    
    print(f'El porcentaje de datos en el rango [{mu-sigma1}, {mu+sigma1}] es {porcentaje_1sigma}%')
    print(f'El porcentaje de datos en el rango [{mu-sigma2}, {mu+sigma2}] es {porcentaje_2sigma}%')
    print(f'El porcentaje de datos en el rango [{mu-sigma3}, {mu+sigma3}] es {porcentaje_3sigma}%')
```

```python
tarea(mu, sigma, 1000)
```
