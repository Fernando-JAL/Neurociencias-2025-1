```python
import scipy.io 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import networkx as nx
```

```python
mat_path = r"fve32.mat"
#archivo que vamos a usar
scipy.io.loadmat(mat_path)
```

```python
def open_mat_file(path: str):
    mat_j = scipy.io.loadmat(mat_path)
    keys = list(mat_j.keys())
    print(keys)
    #hacemos una lista
    lista = []
    for idx in range(3, len(mat_j.keys())):
        lista.append(mat_j[keys[idx]])
    return mat_j, lista
```

```python
j, (names,cij) = open_mat_file(mat_path)
```

```python
cij
```

```python
names
```

```python
#hacer figura
fig = plt.figure(figsize=(8, 8))
ax = sns.heatmap(cij,cmap="magma")
plt.xticks(range(len(cij)), names)
plt.yticks(range(len(cij)), names, rotation=0)
plt.show()
```

```python
G = nx.from_numpy_array(cij)
G.edges(data=True)
```

```python
G.nodes
```

```python
list(G.adjacency())
```

```python
node_lab = {idx: names[idx] for idx in range(31)}
```

```python
nx.draw(G, labels=node_lab, with_labels=True, font_weight='bold')
```

```python
fig = plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
nx.draw(G, labels=node_lab, with_labels=True, font_weight='bold')

plt.subplot(2, 2, 2)
nx.draw_shell(G, labels=node_lab, with_labels=True, font_weight='bold')

plt.subplot(2, 2, 3)
nx.draw_spectral(G, labels=node_lab, with_labels=True, font_weight='bold')

plt.subplot(2, 2, 4)
nx.draw_spring(G, labels=node_lab, with_labels=True, font_weight='bold')
```

```python
nx.draw_circular(G, labels=node_lab, with_labels=True, font_weight='bold')
```

```python
H = nx.relabel_nodes(G, node_lab)

fig = plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
nx.draw(H, with_labels=True, font_weight='bold')

plt.subplot(2, 2, 2)
nx.draw_shell(H, with_labels=True, font_weight='bold')

plt.subplot(2, 2, 3)
nx.draw_spectral(H, with_labels=True, font_weight='bold')

plt.subplot(2, 2, 4)
nx.draw_spring(H, with_labels=True, font_weight='bold')
```
