```python
# 9.- Crear una función que reciba un entero (mu), un flotante (sigma) y un entero (total de números) y: genere array de números con distribución normal, con mu, y sigma. Plotee la distribución, calcule los sigmas, calcule el porcentaje de números que están en los rangos de +-1sigmam +-2sigma, +-3sigma
import seaborn as sns 

mu = 0
sigma = 1
total_numbers = 1000

def normal_distribution(mu, sigma, total_numbers):
    random_array = np.random.normal(mu, sigma, total_numbers)
    plt.figure(figsize=(10, 6))
    sns.histplot(random_array, kde=True, color='blue', bins=30)
    plt.title(f'Distribución Normal: mu={mu}, sigma={sigma}')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.show()
    sigma_values = {
        '1σ': (mu - sigma, mu + sigma),
        '2σ': (mu - 2*sigma, mu + 2*sigma),
        '3σ': (mu - 3*sigma, mu + 3*sigma)
    }
    percentages = {}
    for key, (lower, upper) in sigma_values.items():
        values_in_range = np.sum((random_array >= lower) & (random_array <= upper))
        percentage_in_range = (values_in_range / total_numbers) * 100
        percentages[key] = percentage_in_range
    for key, value in percentages.items():
        print(f"Porcentaje de números en el rango {key}: {value:.2f}%")
    return random_array, percentages

normal_distribution(mu, sigma, total_numbers)
```

```python
# 10.- Usar la función anterior con los valores (0, 1., 1000), (3, 5, 5000), (-3, 3, 10000)
# Usar la función normal_distribution con los valores proporcionados


def normal_distribution(mu, sigma, total_numbers):
    random_array = np.random.normal(mu, sigma, total_numbers)
    
    plt.figure(figsize=(10, 6))
    sns.histplot(random_array, kde=True, color='blue', bins=30)
    plt.title(f'Distribución Normal: mu={mu}, sigma={sigma}')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.show()

    sigma_values = {
        '1σ': (mu - sigma, mu + sigma),
        '2σ': (mu - 2*sigma, mu + 2*sigma),
        '3σ': (mu - 3*sigma, mu + 3*sigma)
    }
    percentages = {}
    for key, (lower, upper) in sigma_values.items():
        values_in_range = np.sum((random_array >= lower) & (random_array <= upper))
        percentage_in_range = (values_in_range / total_numbers) * 100
        percentages[key] = percentage_in_range

    for key, value in percentages.items():
        print(f"Porcentaje de números en el rango {key}: {value:.2f}%")
    
    return random_array, percentages


# Caso 1: (mu=0, sigma=1, total_numbers=1000)
print("Caso 1: mu=0, sigma=1, total_numbers=1000")
normal_distribution(0, 1, 1000)

# Caso 2: (mu=3, sigma=5, total_numbers=5000)
print("\nCaso 2: mu=3, sigma=5, total_numbers=5000")
normal_distribution(3, 5, 5000)

# Caso 3: (mu=-3, sigma=3, total_numbers=10000)
print("\nCaso 3: mu=-3, sigma=3, total_numbers=10000")
normal_distribution(-3, 3, 10000)
```
