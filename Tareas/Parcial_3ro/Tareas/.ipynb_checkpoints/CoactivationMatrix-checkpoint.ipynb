```python
import scipy.io 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import networkx as nx
import pandas as pd
```

```python
mat_path = r"Coactivation_matrix.mat"
scipy.io.loadmat(mat_path)
```

```python
mat_json = scipy.io.loadmat(mat_path)
mat_json
```

```python
mat_dict = {k:v for k, v in mat_json.items() if k[0] != '_'}
mat_dict
```

```python
df = pd.DataFrame(mat_dict['Coactivation_matrix'],)
df
```

```python
fig = plt.figure(figsize=(15, 15))

plt.imshow(mat_dict['Coactivation_matrix'], cmap='hot', interpolation='nearest')
plt.show()
```

```python
fig = plt.figure(figsize=(15, 15))
ax = sns.heatmap(mat_dict['Coactivation_matrix'], cmap='viridis')
plt.show()
```

```python
G = nx.from_pandas_adjacency(df, create_using=nx.DiGraph)

graph_list = list(G.adjacency())

grads = {}
for idx in range(len(graph_list)):
    grads[graph_list[idx][0]] = len(graph_list[idx][1])
```
