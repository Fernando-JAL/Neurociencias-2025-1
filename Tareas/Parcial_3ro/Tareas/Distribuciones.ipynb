```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(5002)

mu = np.random.randint(-3, 3)  #[-3, 3]
sigma = abs(np.random.uniform(-10, 10)) 

data = np.random.normal(loc=mu, scale=sigma, size=1000) #media(mu) y desviación estandar(sigma)
  
```

```python
plt.hist(data, bins=30, density=True, color='purple')
plt.title("Distribución random")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
sigma1 = sigma * 1 #(una desviación estandar) 
sigma2 = sigma * 2 #(media por 2 = 3 desiviaciónes)
sigma3 = sigma * 3 #(media por 3 = 3 desiviaciónes) 

print(sigma1,sigma2,sigma3)
```

```python
porcentajesigma1 = sum(((mu - sigma1)< data) & ((mu + sigma1)>data)) / len(data) * 100
porcentajesigma2 = sum(((mu - sigma2)< data) & ((mu + sigma2)>data)) / len(data) * 100
porcentajesigma3 = sum(((mu - sigma3)< data) & ((mu + sigma3)>data)) / len(data) * 100

porcentajesigma1
```

```python
# Función 

def analisis_distribución(mu, sigma, num_total):
    
    sigma = abs(sigma)  
    data = np.random.normal(mu, sigma, num_total)
    
    plt.hist(data, density=True, bins=40, color='purple')
    plt.title("Distribución random")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
    
    sigma_1, sigma_2, sigma_3 = sigma, 2 * sigma, 3 * sigma
    
    print("Sigma 1: ",sigma_1)
    print("Sigma 2: ",sigma_2)
    print("Sigma 3: ",sigma_3)
    
    vol_sigma1 = np.sum((data >= mu - sigma_1) & (data <= mu + sigma_1)) / num_total * 100
    vol_sigma2 = np.sum((data >= mu - sigma_2) & (data <= mu + sigma_2)) / num_total * 100
    vol_sigma3 = np.sum((data >= mu - sigma_3) & (data <= mu + sigma_3)) / num_total * 100

    print("Porcentaje sigma 1: ",vol_sigma1)
    print("Porcentaje sigma 2: ",vol_sigma2)
    print("Porcentaje sigma 3: ",vol_sigma3)
```

```python

result_1 = analisis_distribución(0, 1.0, 1000)
result_2 = analisis_distribución(3, 5, 5000)
result_3 = analisis_distribución(-3, 3, 10000)
```
