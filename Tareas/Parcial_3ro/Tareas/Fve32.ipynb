```python
import scipy.io 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import networkx as nx
```

```python
mat_path = r"fve32.mat"
scipy.io.loadmat(mat_path)
```

```python
def open_mat_file(path: str):
    mat_json = scipy.io.loadmat(mat_path)
    keys = list(mat_json.keys())
    print(keys)
    
    lista = []
    for idx in range(3, len(mat_json.keys())):
        lista.append(mat_json[keys[idx]])
    return mat_json, lista
```

```python
json, (names,cij) = open_mat_file(mat_path)
```

```python
cij
```

```python
names
```

```python
fig = plt.figure(figsize=(8, 8))
ax = sns.heatmap(cij,cmap="flare")
plt.xticks(range(len(cij)), names)
plt.yticks(range(len(cij)), names, rotation=0)
plt.show()
```

```python
G = nx.from_numpy_array(cij)
G.edges(data=True)
```

```python
G.nodes
```

```python
list(G.adjacency())
```

```python
node_labels = {idx: names[idx] for idx in range(31)}
```

```python
nx.draw(G, labels=node_labels, with_labels=True, font_weight='bold',node_color='purple')
```

```python
fig = plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
nx.draw(G, labels=node_labels, with_labels=True, font_weight='bold',node_color='purple')

plt.subplot(2, 2, 2)
nx.draw_shell(G, labels=node_labels, with_labels=True, font_weight='bold',node_color='purple')

plt.subplot(2, 2, 3)
nx.draw_spectral(G, labels=node_labels, with_labels=True, font_weight='bold',node_color='purple')

plt.subplot(2, 2, 4)
nx.draw_spring(G, labels=node_labels, with_labels=True, font_weight='bold',node_color='purple')
```

```python
nx.draw_kamada_kawai(G, labels=node_labels, with_labels=True, font_weight='bold',node_color='purple')
```

```python
H = nx.relabel_nodes(G, node_labels)

fig = plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
nx.draw(H, with_labels=True, font_weight='bold',node_color='purple')

plt.subplot(2, 2, 2)
nx.draw_shell(H, with_labels=True, font_weight='bold',node_color='purple')

plt.subplot(2, 2, 3)
nx.draw_spectral(H, with_labels=True, font_weight='bold',node_color='purple')

plt.subplot(2, 2, 4)
nx.draw_spring(H, with_labels=True, font_weight='bold',node_color='purple')
```
