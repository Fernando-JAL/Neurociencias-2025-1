```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(5002)

mu = np.random.randint(-3, 3)
sigma = abs(np.random.uniform(-10, 10)) 

data = np.random.normal(loc=mu, scale=sigma, size=1000)
```

```python
normal_data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(normal_data, bins=30, density=True, color='pink')
plt.title("Distribución normal")
plt.xlabel("X")
plt.ylabel("Frecuencia")
plt.show()
```

```python
sigma1 = sigma * 1
sigma2 = sigma * 2
sigma3 = sigma * 3

print(sigma1, sigma2, sigma3)
```

```python
porc_sigma1 = sum(((mu - sigma1) < data) & (data < (mu + sigma1))) / len(data) * 100
porc_sigma2 = sum(((mu - sigma2) < data) & (data < (mu + sigma2))) / len(data) * 100
porc_sigma3 = sum(((mu - sigma3) < data) & (data < (mu + sigma3))) / len(data) * 100
```

```python
print(f"Porcentaje dentro de 1 desviación estándar: {porc_sigma1}%")
print(f"Porcentaje dentro de 2 desviaciones estándar: {porc_sigma2}%")
print(f"Porcentaje dentro de 3 desviaciones estándar: {porc_sigma3}%")
```

```python
def analisis_distribución(mu, sigma, num_total):
    
    sigma = abs(sigma)  
    data = np.random.normal(mu, sigma, num_total)
    
    plt.hist(data, density=True, bins=40, color='pink')
    plt.title("Distribución random")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()
    
    sigma_1, sigma_2, sigma_3 = sigma, 2 * sigma, 3 * sigma
    
    print("Sigma 1: ",sigma_1)
    print("Sigma 2: ",sigma_2)
    print("Sigma 3: ",sigma_3)
    
    vol_sigma1 = np.sum((data >= mu - sigma_1) & (data <= mu + sigma_1)) / num_total * 100
    vol_sigma2 = np.sum((data >= mu - sigma_2) & (data <= mu + sigma_2)) / num_total * 100
    vol_sigma3 = np.sum((data >= mu - sigma_3) & (data <= mu + sigma_3)) / num_total * 100

    print("Porcentaje sigma 1: ",vol_sigma1)
    print("Porcentaje sigma 2: ",vol_sigma2)
    print("Porcentaje sigma 3: ",vol_sigma3)
```

```python
result_1 = analisis_distribución(0, 1.0, 1000)
result_2 = analisis_distribución(3, 5, 5000)
result_3 = analisis_distribución(-3, 3, 10000)
```
