# Mundo pequeño
- Definición: En teoría de redes, el concepto se refiere a la observación de que la mayoría de los individuos están conectados a través de una cadena corta de conocidos comunes. 

```python
# De esta forma se va a requerir programar que:
# 1-. Importar libreria
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Crear el Grafo:
#Definir los nodos y las conexiones.
G = nx.Graph() 
G.add_edges_from([(1, 2), (1, 3), (2, 3), (2, 4)])
```

```python
#Analizar el Grafo:
#Calcular métricas como el diámetro, el coeficiente de clustering, etc.

print(nx.diameter(G))
```

```python
#Visualizar el Grafo:
#Usar matplotlib para crear una visualización.
nx.draw(G, with_labels=True)
plt.show()
```

```python
# Ejemplo:
import networkx as nx

# Crear un grafo Watts-Strogatz, un modelo clásico de mundo pequeño
G = nx.watts_strogatz_graph(1000, 20, 0.1)

# Calcular el camino más corto promedio
avg_shortest_path = nx.average_shortest_path_length(G)
print("Camino más corto promedio:", avg_shortest_path)
```

# Rich club
- Definición: Se refiere a un subgrafo donde los nodos de alto grado están más conectados entre sí que lo que se esperaría por azar. 

```python
import networkx as nx

# Crear un grafo aleatorio (ejemplo)
G = nx.barabasi_albert_graph(1000, 5)

# Obtener los nodos con mayor grado
high_degree_nodes = sorted(G.degree, key=lambda x: -x[1])[:100]  # Top 100 nodos
# Crear un subgrafo con los nodos de alto grado
subgraph = G.subgraph(high_degree_nodes)

# Calcular la densidad del subgrafo
density_subgraph = nx.density(subgraph)

# Calcular la densidad promedio de la red completa
density_G = nx.density(G)

# Comparar las densidades
print("Densidad del subgrafo (rico club):", density_subgraph)
print("Densidad de la red completa:", density_G)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un grafo aleatorio (por ejemplo, usando el modelo Barabási-Albert)
G = nx.barabasi_albert_graph(100, 5)  # Grafo con 100 nodos, cada nuevo nodo se conecta a 5 existentes

# Calcular el coeficiente de "rich club" para diferentes valores de grado k
rich_club_coeff = nx.rich_club_coefficient(G, normalized=True)

# Mostrar los resultados
print("Coeficiente de rich club para diferentes grados:")
for k, coeff in rich_club_coeff.items():
    print(f"Grado k={k}: {coeff}")

# Graficar el coeficiente de rich club
plt.figure(figsize=(10, 5))
plt.plot(list(rich_club_coeff.keys()), list(rich_club_coeff.values()), marker='o')
plt.xlabel("Grado mínimo (k)")
plt.ylabel("Coeficiente de Rich Club")
plt.title("Coeficiente de Rich Club en función del grado mínimo")
plt.grid(True)
plt.show()
```
