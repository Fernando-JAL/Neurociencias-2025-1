1-. Crear un dataframe con la información del ndarray de coativation_matrix, las columnas deben ser números desde el 1, los índices deben ser iguales a las columnas


```python
#mat_path = "/Users/ZvalEnes/Desktop/Neurociencias-2025-1/BCT/BCT/2019_03_03_BCT/data_and_demos/Coativation_matrix.mat"
import pandas as pd 
import scipy.io
 
# Ruta al archivo .mat
file_path = '/Users/ZvalEnes/Desktop/Neurociencias-2025-1/BCT/BCT/2019_03_03_BCT/data_and_demos/Coactivation_matrix.mat'

# Cargar el archivo .mat
mat_data = scipy.io.loadmat(file_path)

# Extraer la matriz de coactivación
coactivation_matrix = mat_data['Coactivation_matrix']

# Crear el DataFrame con índices y columnas numeradas desde 1
df = pd.DataFrame(coactivation_matrix, 
                  columns=range(1, coactivation_matrix.shape[1] + 1), 
                  index=range(1, coactivation_matrix.shape[0] + 1))

# Mostrar los primeros valores del DataFrame
df.head()
```

2-. Mostrar el heatmap de coativation_matrix


```python
import seaborn as sns
import matplotlib.pyplot as plt


# Configurar el tamaño del heatmap
plt.figure(figsize=(10, 8))

# Crear el heatmap
sns.heatmap(df, cmap="Set3", cbar=True)
plt.title("Heatmap de Coactivation Matrix")
plt.xlabel("Columnas")
plt.ylabel("Filas")
plt.show()
```

3-. Crear un dataframe subconjunto del dataframe anterior que contenga de la fila 10 a la 30 y de la columna 10 a la 30


```python
# Seleccionando el subconjunto
subconjunto = df.iloc[9:30, 9:30]  # Nota: los índices comienzan desde 0

print(subconjunto)
subconjunto = df.loc['fila_10':'fila_30', 'columna_10':'columna_30']
```

```python
# Create a subset DataFrame from rows 10 to 30 and columns 10 to 30
df_subset = df.loc[10:30, 10:30]

# Display the subset DataFrame
df_subset
```

4-. - Mostrar el heatmap del nuevo dataframe


```python
# Set plot size for the subset heatmap
plt.figure(figsize=(8, 6))

# Generate the heatmap for the subset DataFrame
sns.heatmap(df_subset, cmap='Set3', cbar=True)

plt.title("Mapa de calor del subconjunto de la matriz de coactivación (Filas 10-30, Columnas 10-30)")
plt.xlabel("Regiones (Subconjunto)")
plt.ylabel("Regiones (Subconjunto)")

plt.show()
```

5-. Crear el grafo dirigido del nuevo dataframe


```python
import networkx as nx

# Crear el grafo dirigido a partir del subconjunto del DataFrame
G = nx.from_pandas_adjacency(df_subset, create_using=nx.DiGraph)

# Dibujar el grafo dirigido
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)  # Definir disposición del grafo
nx.draw(G, pos, with_labels=True, node_color='pink', edge_color='green', node_size=700, font_size=10, arrows=True)
plt.title("Grafo Dirigido del Subconjunto de la Matriz de Coactivación (Filas 10-30, Columnas 10-30)")
plt.show()
```

6-. ¿qué nodo tiene más conexiones del nuevo dataset?


```python
# Calcular el grado de entrada y salida para cada nodo en el grafo dirigido
in_degrees = dict(G.in_degree())  # Número de aristas entrantes
out_degrees = dict(G.out_degree())  # Número de aristas salientes

# Encontrar los nodos con el máximo grado de entrada y salida
max_in_degree_node = max(in_degrees, key=in_degrees.get)
max_out_degree_node = max(out_degrees, key=out_degrees.get)

max_in_degree_node, in_degrees[max_in_degree_node], max_out_degree_node, out_degrees[max_out_degree_node]
```

7-. Del dataframe con el array de coactivation_matrix, calcular los quantiles 0.25, 0.5, 0.75 de los valores de los datos (quantiles de las conexiones)


```python
# Calcular los cuantiles 0.25, 0.5 y 0.75 del DataFrame completo
quantiles = df.quantile([0.25, 0.5, 0.75])

print("Cuantiles de las conexiones:")
print(quantiles)
```

8-. Filtre el dataframe usando como threshold los quantiles 0.25 y 0.5, cómo queda la matriz de activación filtrada?


```python
# Calcular los cuantiles 0.25 y 0.5 y tomar sus medias
q25 = df.quantile(0.25).mean()
q50 = df.quantile(0.5).mean()

# Filtrar el DataFrame
df_filtered = df.where((df >= q25) & (df <= q50))

print(df_filtered)
```
