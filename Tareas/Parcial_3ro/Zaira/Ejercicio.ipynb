3-. Genera un array de 1000 números con distribución normal, con mu= ej1, y sigma= ej2


```python
import numpy as np

# Parámetros de la distribución
mu = 0  # Media de la distribución
sigma = 1  # Desviación estándar de la distribución

# Generar el array de 1000 números
array_normal = np.random.normal(mu, sigma, 1000)
array_normal
```

4-. Plotear la distribución de los 1000 números


```python
import matplotlib.pyplot as plt
import numpy as np

# Crear el histograma de los datos
plt.hist(array_normal, bins=30, density=True, alpha=0.6, color='blue')

# Graficar la curva de la distribución normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(- (x - mu)**2 / (2 * sigma**2))
plt.plot(x, p, 'k', linewidth=2)

plt.xlabel("Valor")
plt.ylabel("Densidad de probabilidad")
plt.title("Distribución normal de 1000 números")
plt.show()
```

4.1 -. Calcular media , desv. estandar, varianza


```python
# Suponiendo que array_normal contiene tus datos
mu = np.mean(array_normal)            # Media
sigma = np.std(array_normal)          # Desviación estándar
variance = np.var(array_normal)       # Varianza

print("Media:", mu)
print("Desviación estándar:", sigma)
print("Varianza:", variance)
```

5-. Calcular el porcentaje de números que están en el rango [-1sigma, 1sigma]


```python
# Contar cuántos números están dentro del rango [-1sigma, 1sigma]
count_in_range = np.sum((array_normal >= (mu - sigma)) & (array_normal <= (mu + sigma)))

# Calcular el porcentaje
percentage_in_range = (count_in_range / len(array_normal)) * 100

print("Porcentaje de números en el rango [-1sigma, 1sigma]:", percentage_in_range, "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt


def generar_array(mu, sigma, total_de_numeros):
    return np.random.normal(mu, sigma, total_de_numeros)


def calcular_sigmas(array):
    mu = np.mean(array)
    sigma = np.std(array)
    return mu, sigma


def plotear_distribucion(array, mu, sigma):
    plt.hist(array, bins=30, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f'Distribución Normal con $\\mu={mu}$ y $\\sigma={sigma}$')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.show()


# Ejemplo de uso:
mu = 0
sigma = 1
total_de_numeros = 1000
array_numeros = generar_array(mu, sigma, total_de_numeros)
mu_calculado, sigma_calculado = calcular_sigmas(array_numeros)
plotear_distribucion(array_numeros, mu_calculado, sigma_calculado)
print(f'\\mu calculado: {mu_calculado}, \\sigma calculado: {sigma_calculado}')
```

```python
def calcular_porcentajes(array, mu, sigma): 
    rango1 = np.sum(np.abs(array - mu) <= sigma) / len(array) * 100 
    rango2 = np.sum(np.abs(array - mu) <= 2 * sigma) / len(array) * 100 
    rango3 = np.sum(np.abs(array - mu) <= 3 * sigma) / len(array) * 100 
    return rango1, rango2, rango3
```

9-. Usar la función anterior con los valores (0, 1., 1000), (3, 5, 5000), (-3, 3, 10000)


Valores (0, 1., 1000)


```python
# Conjunto 1
mu1, sigma1, total1 = 0, 1, 1000 
array1 = generar_array(mu1, sigma1, total1) 
mu1_calculado, sigma1_calculado = calcular_sigmas(array1) 
plotear_distribucion(array1, mu1_calculado, sigma1_calculado) 
rango1_1, rango2_1, rango3_1 = calcular_porcentajes(array1, mu1_calculado, sigma1_calculado) 
print(f'Conjunto 1 (mu={mu1}, sigma={sigma1}):') 
print(f'Porcentaje dentro de ±1σ: {rango1_1:.2f}%') 
print(f'Porcentaje dentro de ±2σ: {rango2_1:.2f}%') 
print(f'Porcentaje dentro de ±3σ: {rango3_1:.2f}%')
```
