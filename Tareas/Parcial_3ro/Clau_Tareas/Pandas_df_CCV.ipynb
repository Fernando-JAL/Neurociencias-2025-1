```python
import numpy as np
import matplotlib.pyplot as plt  
```

```python
np.random.seed(5002)

mu = np.random.randint(-3, 3)  # Media
sigma = abs(np.random.uniform(-10, 10)) # Desviación Estándar
data = np.random.normal(loc=mu, scale=sigma, size=1000)
```

```python
plt.hist(data, bins=30, density=True, color='black')
plt.title("Dist. Aleatoria")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
```

```python
s1 = sigma * 1  
s2 = sigma * 2 # Mu * 2 = 3 desv 
s3 = sigma * 3 # Mu* 3 = 3 desv 

print(s1)
print(s2)
print(s3)
```

```python
# Porcentajes
p_s1 = sum(((mu - s1)< data) & ((mu + s1)>data)) / len(data) * 100
p_s2 = sum(((mu - s2)< data) & ((mu + s2)>data)) / len(data) * 100
p_s3 = sum(((mu - s3)< data) & ((mu + s3)>data)) / len(data) * 100

print(p_s1)
print(p_s2)
print(p_s3)
```

```python
def analisis_dist(mu, sigma, num_total):    
    sigma = abs(sigma)  
    data = np.random.normal(mu, sigma, num_total)
    
    plt.hist(data, density=True, bins=40, color='black')
    plt.title("Dist. Aleatoria")
    plt.xlabel("Eje X")
    plt.ylabel("Eje Y")
    plt.show()
    s_1, s_2, s_3 = sigma, 2 * sigma, 3 * sigma

    print("Sigma 1: ", s_1)
    print("Sigma 2: ", s_2)
    print("Sigma 3: ", s_3)

    vol_s1 = np.sum((data >= mu - s_1) & (data <= mu + s_1)) / num_total * 100
    vol_s2 = np.sum((data >= mu - s_2) & (data <= mu + s_2)) / num_total * 100
    vol_s3 = np.sum((data >= mu - s_3) & (data <= mu + s_3)) / num_total * 100

    print("% Sigma 1: ", vol_s1)
    print("% Sigma 2: ",vol_s2)
    print("% Sigma 3: ",vol_s3)
```

```python
# Resultados
r_1 = analisis_dist(0, 1.0, 1000)
```

```python
r_2 = analisis_dist(3, 5, 5000)
```

```python
r_3 = analisis_dist(-3, 3, 10000)
```
