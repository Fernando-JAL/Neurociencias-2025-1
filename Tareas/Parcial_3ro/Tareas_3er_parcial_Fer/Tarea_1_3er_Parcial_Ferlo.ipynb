### Crear un dataframe con la información del ndarray de coativation_matrix, las columnas deben ser números desde el 1, los índices deben ser iguales a las columnas


```python
import scipy.io
import pandas as pd
```

```python
data = scipy.io.loadmat('Coactivation_matrix.mat')
```

```python
print(data.keys())
```

```python
Coactivation_matrix = data['Coactivation_matrix']
```

```python
df = pd.DataFrame(Coactivation_matrix)
df.index += 1
df.columns = df.index
df.head()
```

### Mostrar el heatmap de coativation_matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df, cmap='viridis', annot=False, cbar=True)
plt.title('Heatmap de Coactivation Matrix')
plt.xlabel('Columnas')
plt.ylabel('filas')
plt.show()
```

### Crear un dataframe subconjunto del dataframe anterior que contenga de la fila 10 a la 30 y de la columna 10 a la 30

```python
df_subset = df.iloc[9:30, 9:30]
df_subset
```

### Mostrar el heatmap del nuevo dataframe

```python
plt.figure(figsize=(10,8))
sns.heatmap(df_subset, cmap='viridis', annot=False)
plt.title('Heatmap del subset')
plt.show()
```

### Crear el grafo dirigido del nuevo dataframe

```python
import networkx as nx
```

```python
G = nx.from_pandas_adjacency(df_subset, create_using=nx.DiGraph)
```

```python
plt.figure(figsize=(12,10))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color="skyblue", font_size=10, font_weight="bold", edge_color="gray", arrows=True)
plt.title('Grafo dirigido del subset')
plt.show()
```

### ¿Qué nodo tiene más conexiones del nuevo dataset?

```python
in_degrees = dict(G.in_degree())
out_degrees = dict(G.out_degree())

max_in_degree_node = max(in_degrees, key=in_degrees.get)
max_in_degree_value = in_degrees[max_in_degree_node]

max_out_degree_node = max(out_degrees, key=out_degrees.get)
max_out_degree_value = out_degrees[max_out_degree_node]

print(f"Nodo con más conexiones de entrada: {max_in_degree_node} (Conexiones: {max_in_degree_value})")
print(f"Nodo con más conexiones de salida: {max_out_degree_node} (Conexiones: {max_out_degree_value})")
```

### Del dataframe con el array de coactivation_matrix, calcular los quantiles 0.25, 0.5, 0.75 de los valores de los datos (quantiles de las conexiones)

```python
quantiles_all = df.values.flatten()
quantiles_result = pd.Series(quantiles_all).quantile([0.25, 0.5, 0.75])

print("Cuantiles generales de todas las conexiones:")
print(quantiles_result)
```

### Filtre el dataframe usando como threshold los quantiles 0.25 y 0.5, cómo queda la matriz de activación filtrada?

```python
import numpy as np

quantile_25 = np.quantile(df.values.flatten(), 0.25)
quantile_50 = np.quantile(df.values.flatten(), 0.5)

activation_matrix_filtered = df[(df >= quantile_25) & (df <= quantile_50)].fillna(0)

print("Matriz de activación filtrada:")
print(activation_matrix_filtered)
```
