## 1. Ejercicios de grafos
Del .mat file de coativation_matrix, realizar los siguientes ejercicios:

1. 
Crear un dataframe con la información del ndarray de coativation_matrix, las columnas deben ser números desde el 1, los índices deben ser iguales a las columnaores a 0.

```python
coactivation_matrix = coact_matrix['Coactivation_matrix']

num_col = coactivation_matrix.shape[1]
columns = range(1, num_col + 1)
df_cm = pd.DataFrame(coactivation_matrix, index=columns, columns=columns)

df_cm
```

2. Mostrar el heatmap de coativation_matrix


```python
sns.heatmap(df_cm, cmap='coolwarm')
plt.title("Heatmap de coactivation_matrix")
```

3. Crear un dataframe subconjunto del dataframe anterior que contenga de la fila 10 a la 30 y de la columna 10 a la 30


```python
df_sub = df_cm.iloc[9:30, 9:30]
df_sub
```

4. Mostrar el heatmap del nuevo dataframe


```python
sns.heatmap(df_sub, cmap='coolwarm')
plt.title("Heatmap de coactivation_matrix (subconjunto)")
```

5. Crear el grafo dirigido del nuevo dataframe


```python
G = nx.from_pandas_adjacency(df_sub, create_using=nx.DiGraph)
nx.draw_networkx(G, arrows=True, arrowstyle='-|>')
plt.title("Grafo dirigido del subconjunto")
```

6. ¿Qué nodo tiene más conexiones del nuevo dataset?


```python
node_degrees = G.degree() 
max_degree_node = max(node_degrees, key=lambda x: x[1])
max_degree_node
```
