9. Crear una función que reciba un entero (mu), un flotante (sigma) y un entero (total de números) y: genere array de números con distribución normal, con mu, y sigma. Plotee la distribución, calcule los sigmas, calcule el porcentaje de números que están en los rangos de +-1sigmam +-2sigma, +-3sigma

```python
def f_array ():
    mu = int(input("Ingresa un número entero: "))
    sigma = float(input("Ingresa un flotante: "))
    total_num = int(input("Ingresa un número entero que será el total de elementos de tu array: "))
    
    array_usuario = np.random.normal(mu, sigma, total_num)
    sns.histplot(array_usuario, kde=True, color='hotpink')
    plt.title('Distribución normal de los números')
    plt.xlabel('Valores')
    plt.ylabel('Frecuencia')
    
    sigma1 = sigma 
    sigma2 = sigma*2
    sigma3 = sigma*3
    sigma1_neg = sigma * -1
    sigma2_neg = sigma * -2
    sigma3_neg = sigma * -3
    
    porcentaje1 = np.sum((array_usuario >= sigma1_neg) & (array_usuario <= sigma))/ len(array_usuario)*100
    porcentaje2 = np.sum((array_usuario >= sigma2_neg) & (array_usuario <= sigma2))/ len(array_usuario)*100
    porcentaje3 = np.sum((array_usuario >= sigma3_neg) & (array_usuario <= sigma3))/ len(array_usuario)*100
    
    print(porcentaje1)
    print(porcentaje2)
    print(porcentaje3)
```
