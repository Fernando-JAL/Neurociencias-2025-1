## Ejercicio 10

```python
def analizar_distribucion(mu, sigma, total):
    # Generar un array de números con distribución normal
    array_normal = np.random.normal(mu, sigma, total)

    # Graficar la distribución con Seaborn
    plt.figure(figsize=(10, 6))
    sns.histplot(array_normal, bins=30, kde=True, color='r', stat='density', label='Histograma')
    plt.title(f'Distribución Normal (mu={mu}, sigma={sigma})')
    plt.xlabel('Valor')
    plt.ylabel('Densidad de probabilidad')
    plt.legend()

    # Mostrar el gráfico
    plt.show()

    # Calcular los rangos de 1σ, 2σ, 3σ
    sigma_1_range = (mu - sigma, mu + sigma)
    sigma_2_range = (mu - 2*sigma, mu + 2*sigma)
    sigma_3_range = (mu - 3*sigma, mu + 3*sigma)

    # Calcular el porcentaje de valores dentro de los rangos
    in_1sigma = np.mean((array_normal >= sigma_1_range[0]) & (array_normal <= sigma_1_range[1])) * 100
    in_2sigma = np.mean((array_normal >= sigma_2_range[0]) & (array_normal <= sigma_2_range[1])) * 100
    in_3sigma = np.mean((array_normal >= sigma_3_range[0]) & (array_normal <= sigma_3_range[1])) * 100

    # Mostrar los resultados
    print(f"Porcentaje de valores dentro de [-1σ, 1σ]: {in_1sigma:.2f}%")
    print(f"Porcentaje de valores dentro de [-2σ, 2σ]: {in_2sigma:.2f}%")
    print(f"Porcentaje de valores dentro de [-3σ, 3σ]: {in_3sigma:.2f}%")
```

```python
analizar_distribucion(mu=0, sigma=1.0, total=1000)
```

```python
analizar_distribucion(mu=3, sigma=5.0, total=5000)
```

```python
analizar_distribucion(mu=-3, sigma=3.0, total=10000)
```
