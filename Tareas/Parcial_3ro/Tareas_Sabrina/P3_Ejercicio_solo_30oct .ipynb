```python
import random 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generar un número aleatorio entero en el rango [-3, 3] con seed=5002
random.seed(5002)
mu = random.randint(-3, 3)
print(mu)
```

```python
# Generar un número aleatorio flotante en el rango [-10, 10] con seed=5002
random.seed(5002)
sigma = random.uniform(-10,10)
print(sigma)
```

```python
# Genera un array de 1000 números con distribución normal, con mu= ej1, y sigma= ej2
mu = random.randint(-3, 3)
sigma = random.uniform(-10,10)

array1 = np.random.normal(mu, sigma, 1000)
print(array1)
```

```python
# Plotear la distribución de los 1000 números
sns.histplot(array1, kde=True)
plt.title('Distribución normal de los 1000 números')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
```
