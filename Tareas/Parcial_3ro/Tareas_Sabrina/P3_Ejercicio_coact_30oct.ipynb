```python
import scipy.io
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
```

```python
path = r"/Users/sabrina/Documents/UNI/A-NEUROCIENCIAS/QUINTO /MODELOS/Neurociencias-2025-1/BCT/BCT/2019_03_03_BCT/data_and_demos/Coactivation_matrix.mat"
mat = scipy.io.loadmat(path)
print(mat)
```

```python
# Crear un dataframe con la información del ndarray de coativation_matrix, las columnas deben ser números desde el 1, los índices deben ser iguales a las columnas

coactivation_matrix = mat['Coactivation_matrix']

num_col = coactivation_matrix.shape[1]
columns = range(1, num_col + 1)
df_cm = pd.DataFrame(coactivation_matrix, index=columns, columns=columns)

df_cm
```

```python
# Mostrar el heatmap de coativation_matrix
cijcoact = mat['Coactivation_matrix']
sns.heatmap(cijcoact)
```

```python
# Crear un dataframe subconjunto del dataframe anterior que contenga de la fila 10 a la 30 y de la columna 10 a la 30
df_sub = df_cm.iloc[9:31, 9:31]
print(df_sub)
```

```python
# Mostrar el heatmap del nuevo dataframe
cijcoact_nuevo = df_sub
sns.heatmap(cijcoact_nuevo)
```

```python
# Crear el grafo dirigido del nuevo dataframe
array_df_sub = df_sub.values
G = nx.from_numpy_matrix(array_df_sub, create_using=nx.DiGraph)

# dibujarlo 
nx.draw_networkx(G, with_labels=True, node_color="purple", node_size=300, edge_color="gray", arrows=True)
plt.show()
```

```python
# ¿qué nodo tiene más conexiones del nuevo dataset?
in_degree = dict(G.in_degree())   
out_degree = dict(G.out_degree()) 
total_degrees = dict(G.degree())   
nodo_mas_conexiones = max(total_degrees, key=total_degrees.get)
print(nodo_mas_conexiones)
```

```python
# Del dataframe con el array de coactivation_matrix, calcular los quantiles 0.25, 0.5, 0.75 de los valores de los datos (quantiles de las conexiones)
q1 = df_cm.quantile(0.25)
print(q1)
q2 = df_cm.quantile(0.5)
print(q2)
q3 = df_cm.quantile(0.75)
print(q3)
```

```python
# histograma 
# lo aplanamos con flatten
plt.hist(df_cm.values.flatten(), bins=5, edgecolor='black')
plt.suptitle('Histograma coactivation matrix')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Filtre el nuevo dataframe usando como threshold el valor de 0, es decir, binarizar la matriz de coactivación con valores mayores a 0.
df_sub_nuevo_nuevo = (df_sub > 0).astype(int) 
print(df_sub_nuevo_nuevo)
```
