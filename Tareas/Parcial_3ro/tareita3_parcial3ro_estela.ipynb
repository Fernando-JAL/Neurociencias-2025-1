```python
# Genera un array de 1000 números con distribución normal
# Con mu= ej1, y sigma= ej2

mu = 2
sigma = 2.756897507517177
array = np.random.normal(mu, sigma, 1000)
array
```

```python
# Plotear la distribución de los 1000 números
plt.hist(array, bins = 50, density = True, color='magenta')
plt.title('distribución normal')
plt.show()
```

```python
# Calcular el valor de 1sigma, 2sigma, 3sigma
mean = np.mean(array) # media
std_dev = np.std(array) # desviación estándar

one_sigma = (mean - std_dev, mean + std_dev)
two_sigma = (mean - 2 * std_dev, mean + 2 * std_dev)
three_sigma = (mean - 3 * std_dev, mean + 3 * std_dev)
print('1sigma:', one_sigma)
print('2sigma:', two_sigma)
print('3sigma:', three_sigma)
```

```python
# Calcular el porcentaje de números que están en el rango [-1sigma, 1sigma]
lim_min = mean - std_dev
lim_max = mean + std_dev
conteo = np.sum((array >= lim_min) & (array <= lim_max))
porcentaje_rango1 = (conteo / len(array)) * 100
porcentaje_rango1
```

```python
# Calcular el porcentaje de números que están en el rango [-2sigma, 2sigma]
lim_min = mu - 2 * sigma
lim_max = mu + 2 * sigma
conteo = np.sum((array >= lim_min) & (array <= lim_max))
porcentaje_rango2 = (conteo / len(array)) * 100
porcentaje_rango2
```

```python
# Calcular el porcentaje de números que están en el rango [-3sigma, 3sigma]
lim_min = mu - 3 * sigma
lim_max = mu + 3 * sigma
conteo = np.sum((array >= lim_min) & (array <= lim_max))
porcentaje_rango3 = (conteo / len(array)) * 100
porcentaje_rango3
```
