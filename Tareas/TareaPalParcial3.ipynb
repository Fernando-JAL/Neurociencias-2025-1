Crear el grafo dirigido del nuevo dataframe

```python
import networkx as nx
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
path = r"C:\Users\gia19\OneDrive\Documentos\RepositorioCacahuate\Parcial 2\Coactivation_matrix.mat"
mat_json =scipy.io.loadmat(path)

df= pd.DataFrame(mat_json["Coactivation_matrix"])
df2=df.loc[10:30,10:30]

df_grafo = pd.DataFrame(df2)
G_grafo = nx.from_pandas_adjacency(df_grafo, create_using=nx.DiGraph())

plt.figure(figsize = (12,10))
pos = nx.spring_layout(G_grafo)
nx.draw_networkx(G_grafo, pos=pos, with_labels=True)

plt.title('Grafo dirigido 10:30')
plt.show()
```

¿qué nodo tiene más conexiones del nuevo dataset?


```python
G = G_grafo.copy()
nodo_mas_conectado = max(G.degree, key=lambda x: x[1])[0]

grafo_nodo = G.subgraph([nodo_mas_conectado] + list(G.successors(nodo_mas_conectado)) + list(G.predecessors(nodo_mas_conectado)))

pos = nx.spring_layout(grafo_nodo)

color_nodo = ['red' if node == nodo_mas_conectado else 'blue' for node in grafo_nodo]

nx.draw_networkx(grafo_nodo, pos=pos, with_labels=True, node_color=color_nodo)

plt.title('Subgrafo: nodo más conectado en 10:30')
plt.show()
```

9. Crear una función que reciba un entero (mu), un flotante (sigma) y un entero (total de números) y: genere array de números con distribución normal, con mu, y sigma. Plotee la distribución, calcule los sigmas, calcule el porcentaje de números que están en los rangos de +-1sigmam +-2sigma, +-3sigma

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_distribucion_y_calcular(mu, sigma, total_numeros):
    numeros = np.random.normal(mu, sigma, total_numeros)
    rango_1sigma, rango_2sigma, rango_3sigma = mu + np.array([-1, 1]) * sigma, mu + np.array([-2, 2]) * sigma, mu + np.array([-3, 3]) * sigma
    porcentajes = [
        np.mean((numeros >= r[0]) & (numeros <= r[1])) * 100 
        for r in [rango_1sigma, rango_2sigma, rango_3sigma]
    ]

    plt.hist(numeros, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')
    x = np.linspace(*plt.xlim(), 100)
    plt.plot(x, 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((x - mu) / sigma)**2), 'k')
    for r, color in zip([mu, *rango_1sigma, *rango_2sigma, *rango_3sigma], ['red', 'blue', 'blue', 'green', 'green', 'purple', 'purple']):
        plt.axvline(r, color=color, linestyle='--', linewidth=1)
    plt.show()

    return {'Porcentaje en ±1σ': porcentajes[0], 'Porcentaje en ±2σ': porcentajes[1], 'Porcentaje en ±3σ': porcentajes[2]}


resultados = generar_distribucion_y_calcular(mu=0, sigma=1, total_numeros=1000)
print(resultados)
```

10. Usar la función anterior con los valores (0, 1., 1000), (3, 5, 5000), (-3, 3, 10000)

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_distribucion_y_calcular(mu, sigma, total_numeros):
    numeros = np.random.normal(mu, sigma, total_numeros)
    rangos = [mu + np.array([-k, k]) * sigma for k in range(1, 4)]
    porcentajes = [np.mean((numeros >= r[0]) & (numeros <= r[1])) * 100 for r in rangos]

    plt.hist(numeros, bins=30, density=True, alpha=0.6, color='skyblue')
    plt.plot(np.linspace(*plt.xlim(), 100), 
             1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((np.linspace(*plt.xlim(), 100) - mu) / sigma)**2), 'k')
    for r in rangos: 
        plt.axvline(r[0], color='blue', linestyle='--', linewidth=1)
        plt.axvline(r[1], color='blue', linestyle='--', linewidth=1)
    plt.show()

    return {'±1σ': porcentajes[0], '±2σ': porcentajes[1], '±3σ': porcentajes[2]}

# Pruebas
print(generar_distribucion_y_calcular(0, 1, 1000))
print(generar_distribucion_y_calcular(3, 5, 5000))
print(generar_distribucion_y_calcular(-3, 3, 10000))
```
