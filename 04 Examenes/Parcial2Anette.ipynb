1. Crear una función llamada "df_ordered" que reciba un dataframe y un str y retorne un dataframe ordenado de mayor a menor considerando la característica str

```python
def df_ordered(dataframe, nombre_columna):
    if nombre_columna in dataframe.columns:
        return dataframe.sort_values(by=nombre_columna, ascending=False)
    else:
        return f"La columna '{nombre_columna}' no existe en el dataframe."
    
#Ejemplo
df= mental_ilnesses_prevalence
df_ordered_result = df_ordered(df,'Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized')
df_ordered_result
```

2. Crear una función llamada "plot_bar" que reciba un dataframe y un str y haga el plot de barras de ese dataframe considerando esa caracterísitca descrita por el string

```python
def plot_bar(dataframe, nombre_columna):
    if nombre_columna in dataframe.columns:
        plt.figure(figsize=(10, 6))
        plt.bar(dataframe['Entity'], dataframe[nombre_columna], color='skyblue')
        plt.xlabel('Pais')
        plt.ylabel(nombre_columna)
        plt.title(f' Grafico de barras de: {nombre_columna}')
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
    else:
        print(f"La columna '{nombre_columna}' no existe en el dataframe.")
        
#Ejemplo
df=mental_ilnesses_prevalence
plot_bar(df, 'Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized')     
```

12. mostrar en un mapa de color la correlación entre las features ["Schizophrenia disorders", "Depressive disorders", "Anxiety disorders", "Bipolar disorders", "Eating disorders"]

15. aquellos que tienen dependencia lineal encontrar los valores de la recta. 

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
csv_path= r"C:\Users\gia19\OneDrive\Documentos\RepositorioProfe\cancer_reg.csv"

df_cancer = pd.read_csv(csv_path)

features_dependencia_lineal = ['avganncount', 'avgdeathsperyear']

ft_dependientes_lineales = df_cancer[features_dependencia_lineal]
corr = ft_dependientes_lineales.corr()
corr

df= pd.DataFrame(ft_dependientes_lineales)
df.to_csv("dependientes_lineales")

# Cargar los datos desde un archivo CSV
df = pd.read_csv('dependientes_lineales')

# Asignar las columnas correspondientes
x = df['avganncount']
y = df['avgdeathsperyear']

# Calcular la recta de mejor ajuste (regresión lineal)
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Mostrar los resultados de la regresión
print(f"Pendiente (slope): {slope}")
print(f"Intersección (intercept): {intercept}")
print(f"Coeficiente de correlación (r_value): {r_value}")
print(f"Error estándar (std_err): {std_err}")

# Generar los valores de y de acuerdo con la recta de regresión
y_pred = slope * x + intercept

# Graficar los datos y la recta de mejor ajuste
plt.scatter(x, y, label='Datos')
plt.plot(x, y_pred, color='red', label=f'Recta de mejor ajuste: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.title('Recta de regresión lineal')
plt.legend()
plt.grid(True)
plt.show()
```
