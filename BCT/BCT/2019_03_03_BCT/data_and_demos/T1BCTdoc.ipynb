```python
import scipy.io
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Extraer las matrices CIJall y CIJctx
CIJall = mat_json['CIJall']
CIJctx = mat_json['CIJctx']

# Calcular máximos y mínimos
max_CIJall = np.max(CIJall)
min_CIJall = np.min(CIJall)
max_CIJctx = np.max(CIJctx)
min_CIJctx = np.min(CIJctx)

print(f"Máximos de CIJall: Máx={max_CIJall}, Mín={min_CIJall}")
print(f"Máximos y mínimos de CIJctx: Máx={max_CIJctx}, Mín={min_CIJctx}")
```

```python

# Crear mapa de calor para CIJall
plt.figure(figsize=(10,8))
sns.heatmap(CIJall, cmap='viridis')
plt.title('Mapa de calor CIJall')
plt.show()

# Crear mapa de calor para CIJctx
plt.figure(figsize=(10,8))
sns.heatmap(CIJctx, cmap='jet')
plt.title('Mapa de calor CIJctx')
plt.show()
```
