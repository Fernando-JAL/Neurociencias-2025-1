```python
import scipy.io
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import networkx as nx
```

```python
mat_path = r"../../BCT/BCT/2019_03_03_BCT/data_and_demos/Coactivation_matrix.mat"

mat_json = scipy.io.loadmat(mat_path)
```

```python
mat_json
```

```python
mat_json['Coactivation_matrix'].shape
```

```python
df_subset = mat_json['Coactivation_matrix'][:20, :20]
```

```python
G = nx.from_numpy_array(df_subset, create_using=nx.DiGraph())
```

```python
nx.draw_networkx(G, arrows=True, arrowstyle='-|>')
```

```python
nx.draw_circular(G, with_labels=True)
```

```python
pos = mat_json['Coord'][:20][:, :2]

nx.draw_networkx(G, pos, arrows=True, arrowstyle='-|>')
# edge weight labels
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
```
