# Controladores de flujo

Se trata de una estructura de control condicional. Nos permite evaluar si una o más condiciones se cumplen, para decir qué acción vamos a ejecutar. Esta evaluación de condiciones sólo puede dar un resultado Verdadero o un resultado Falso.


# Sentencia If (Si)
Permite dividir el flujo de un programa en diferentes caminos. El if se ejecuta siempre que la expresión que comprueba devuelva True

```python
if True:
    print('Hola If')
```

```python
if True:
    print("Se cumple la condición")
    print("También se muestre este print")
```

```python
if False:
    print('No va a entrar')
```

```python
a = -2
if a > 0:
    print('a es mayor que 0')
```

```python
a = input()
if int(a) > 0:
    print('a es mayor que 0')
```

Podemos encadenas diferentes If

```python
a = 5
if a == 2:
    print("a vale 2")
if a == 5:
    print("a vale 5")
```

O también anidar If dentro de If

```python
a = 5
b = 10
if a == 5:
    print("a vale: ", a)
    if b == 10:
        print("y b vale: ", b)
```

Como condición podemos evaluar múltiples expresiones, siempre que éstas devuelvan True o False
es decir, podemos usar expresiones anidadas

```python
a = 5
b = 10
if b == 10 and a == 5:
    print("a vale 5 y b vale 10")
```

# Sentencia Else (Sino)
Se encadena a un If para comprobar el caso contrario (en el que no se cumple la condición).

```python
a = 15
if a == 5:
    print('a es igual a 5')
else:
    print('a no es igual a 5')
```

```python
n = 78
if n % 2 == 0:
    print(n, "es un número par")
else:
    print(n, "es un número impar")
```

# Sentencia Elif (Sino Si)
Se encadena a un if u otro elif para comprobar múltiples condiciones, siempre que las anteriores no se ejecuten.

```python
comando = "SALUDAR"
if comando == "ENTRAR":
    print("Bienvenido al sistema")
elif comando == "SALUDAR":
    print("Hola, espero que te lo estés pasando bien aprendiendo Python")
elif comando == "SALIR":
    print("Saliendo del sistema...")
else:
    print("Este comando no se reconoce")
```

```python
a = 20
if a >= 25:
    print('a es >= 25')
elif a >= 15:
    print('a es >= 15')
```

```python
a = int(input())

if a % 2 == 0:
    print('a es par')
if a > 20:
    print('a > 20')
```

```python
nota = float(input("Introduce una nota: "))
if nota >= 7:
    print("Notable")
elif nota >= 6:
    print("Bien")
elif nota >= 5:
    print("Suficiente")
elif nota >= 9:
    print("Sobresaliente")
else:
    print("Insuficiente")
```

```python
nota = float(input("Introduce una nota: "))
if nota >= 9:
    print("Sobresaliente")
elif nota >= 7:
    print("Notable")
elif nota >= 6:
    print("Bien")
elif nota >= 5:
    print("Suficiente")
else:
    print("Insuficiente")
```

Es posible simular el funcionamiento de elif con if utilizando expresiones condicionales

```python
nota = float(input("Introduce una nota: "))
if nota >= 9:
    print("Sobresaliente")
if nota >= 7 and nota < 9:
    print("Notable")
if nota >= 6 and nota < 7:
    print("Bien")
if nota >= 5 and nota < 6:
    print("Suficiente")
if nota < 5:
    print("Insuficiente")
```

# Instrucción pass
Sirve para finalizar un bloque, se puede utilizar en un bloque vacío.

```python
expresion = False
if expresion:
    pass
else:
    print('no ejecutó el pass')
```
